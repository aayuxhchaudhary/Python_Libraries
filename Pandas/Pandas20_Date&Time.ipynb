```python
 # PART 4: EXTRACTING DATE COMPONENTS
print("="*70)
print("EXTRACTING DATE COMPONENTS")
print("="*70)

# Create sample datetime data
datetime_data = pd.DataFrame({
    'Transaction_Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Amount': [100, 150, 200, 125, 175, 300, 250, 225, 180, 210]
})

print("Sample Data:")
print(datetime_data)
```

```python
# Extract various components
print("\n Extracting Date Components:")
datetime_data['Year'] = datetime_data['Transaction_Date'].dt.year
datetime_data['Month'] = datetime_data['Transaction_Date'].dt.month
datetime_data['Month_Name'] = datetime_data['Transaction_Date'].dt.month_name()
datetime_data['Day'] = datetime_data['Transaction_Date'].dt.day
datetime_data['Day_Name'] = datetime_data['Transaction_Date'].dt.day_name()
datetime_data['Day_of_Week'] = datetime_data['Transaction_Date'].dt.dayofweek  # Monday=0, Sunday=6
datetime_data['Weekend'] = datetime_data['Transaction_Date'].dt.dayofweek >= 5
datetime_data['Quarter'] = datetime_data['Transaction_Date'].dt.quarter
datetime_data['Year_Day'] = datetime_data['Transaction_Date'].dt.dayofyear
datetime_data['Week_of_Year'] = datetime_data['Transaction_Date'].dt.isocalendar().week

print(datetime_data.head(10))
```

```python
# Extract time components (if time exists)
print("\n Extracting Time Components:")
datetime_data['Time'] = pd.Timestamp.now()
datetime_data['Hour'] = datetime_data['Time'].dt.hour
datetime_data['Minute'] = datetime_data['Time'].dt.minute
datetime_data['Second'] = datetime_data['Time'].dt.second

print(datetime_data[['Time', 'Hour', 'Minute', 'Second']].head())
```

```python
# PART 5: DATE ARITHMETIC AND CALCULATIONS
# 5.1 Adding and Subtracting Time

print("="*70)
print(" DATE ARITHMETIC")
print("="*70)

# Create base dates
dates = pd.DataFrame({
    'Start_Date': pd.date_range(start='2024-03-01', periods=5, freq='D'),
    'Project': ['Project A', 'Project B', 'Project C', 'Project D', 'Project E']
})

print("Base Data:")
print(dates)
```

```python
# Adding time
print("\n Adding Time to Dates:")
dates['Plus_1_Day'] = dates['Start_Date'] + pd.Timedelta(days=1)
dates['Plus_1_Week'] = dates['Start_Date'] + pd.Timedelta(weeks=1)
dates['Plus_1_Month'] = dates['Start_Date'] + pd.DateOffset(months=1)
dates['Plus_10_Days'] = dates['Start_Date'] + pd.Timedelta(days=10)

print(dates[['Start_Date', 'Plus_1_Day', 'Plus_1_Week', 'Plus_1_Month', 'Plus_10_Days']])
```

```python
# Subtracting time
print("\n Subtracting Time from Dates:")
dates['Minus_1_Day'] = dates['Start_Date'] - pd.Timedelta(days=1)
dates['Minus_1_Week'] = dates['Start_Date'] - pd.Timedelta(weeks=1)
print(dates[['Start_Date', 'Minus_1_Day', 'Minus_1_Week']])
```

```python
# 5.2 Calculating Differences Between Dates
print("="*70)
print("CALCULATING DATE DIFFERENCES")
print("="*70)

# Create project timeline data
timeline = pd.DataFrame({
    'Project': ['Website', 'App', 'Database', 'API', 'Testing'],
    'Start_Date': pd.to_datetime(['2023-01-15', '2024-02-01', '2024-01-20', '2024-02-10', '2024-03-01']),
    'End_Date': pd.to_datetime(['2024-02-20', '2024-03-15', '2024-02-10', '2024-03-20', '2024-03-25'])
})

print(" Project Timeline:")
print(timeline)
```

```python
# Calculate differences
print("\n Calculating Project Durations:")
timeline['Duration_Days'] = (timeline['End_Date'] - timeline['Start_Date']).dt.days
timeline['Duration_Weeks'] = timeline['Duration_Days'] / 7
timeline['Duration_Months'] = timeline['Duration_Days'] / 30.44

print(timeline[['Project', 'Start_Date', 'End_Date', 'Duration_Days', 'Duration_Weeks']])
```

```python
# Calculate age from birthdate
print("\n Calculate Age from Birthdate:")
people = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Birthdate': pd.to_datetime(['1990-05-15', '1985-08-20', '1995-03-10', '1988-12-05'])
})
today = pd.Timestamp.now()
people['Age_Days'] = (today - people['Birthdate']).dt.days
people['Age_Years'] = people['Age_Days'] / 365.25

print(people[['Name', 'Birthdate', 'Age_Years']])
```

```python
# PART 6: FILTERING BY DATES
print("="*70)
print(" FILTERING BY DATES")
print("="*70)

# Create sales data
sales = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', end='2024-03-31', freq='D'),
    'Sales': np.random.randint(1000, 5000, 91)
})

print(" Sales Data (First 10 rows):")
print(sales.head(10))
```

```python
# Filter by specific date
print("\n Filter by specific date:")
specific_date = sales[sales['Date'] == '2024-01-15']
print(specific_date)
```

```python
# Filter by date range
print("\n Filter by date range (January 2024):")
jan_sales = sales[(sales['Date'] >= '2024-01-01') & (sales['Date'] <= '2024-01-31')]
print(f"January sales records: {len(jan_sales)}")
print(jan_sales.head())
```

```python
# Using between() method
print("\n Using between() method:")
feb_sales = sales[sales['Date'].between('2024-02-01', '2024-02-29')]
print(f"February sales records: {len(feb_sales)}")

# Filter by year/month
print("\n Filter by specific month (March):")
sales['Month'] = sales['Date'].dt.month
march_sales = sales[sales['Month'] == 3]
print(f"March sales records: {len(march_sales)}")

# Filter by day of week
print("\n Filter by weekends only:")
sales['Day_Name'] = sales['Date'].dt.day_name()
weekend_sales = sales[sales['Day_Name'].isin(['Saturday', 'Sunday'])]
print(f"Weekend sales records: {len(weekend_sales)}")
print(weekend_sales.head())
```

```python
# PART 7: RESAMPLING TIME SERIES DATA

print("="*70)
print("RESAMPLING TIME SERIES DATA")
print("="*70)

# Create daily sales data
daily_sales = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', end='2024-03-31', freq='D'),
    'Sales': np.random.randint(1000, 5000, 91)
})
daily_sales.set_index('Date', inplace=True)

print("Daily Sales Data (First 10 days):")
print(daily_sales.head(10))
```

```python
# Resample to weekly
print("\n Weekly Sales (Sum):")
weekly_sales = daily_sales.resample('W').sum()
print(weekly_sales.head())

print("\n Weekly Sales (Average):")
weekly_avg = daily_sales.resample('W').mean()
print(weekly_avg.head())
```

```python
# Resample to monthly
print("\n Monthly Sales (Sum):")
monthly_sales = daily_sales.resample('M').sum()
print(monthly_sales)

print("\n Monthly Sales Statistics:")
monthly_stats = daily_sales.resample('M').agg(['sum', 'mean', 'min', 'max', 'std'])
print(monthly_stats)
```

```python
# Different aggregation methods
print("\n Various Aggregations:")
resampled = daily_sales.resample('W').agg({
    'Sales': ['sum', 'mean', 'count', 'min', 'max']
})
print(resampled.head())

# Resample with custom function
print("\n Custom aggregation (first and last day sales):")
weekly_first_last = daily_sales.resample('W').agg(lambda x: x.iloc[0] - x.iloc[-1])
print(weekly_first_last.head())
```

```python
# PART 8: HANDLING TIME ZONES
print("="*70)
print("HANDLING TIME ZONES")
print("="*70)

# Create datetime without timezone
dates_no_tz = pd.date_range(start='2024-03-15 09:00', periods=5, freq='H')
df_tz = pd.DataFrame({'Timestamp': dates_no_tz, 'Value': range(5)})

print("Datetime without timezone:")
print(df_tz)
```

```python
# Add timezone
print("\n Adding timezone (US/Eastern):")
df_tz['US_Eastern'] = df_tz['Timestamp'].dt.tz_localize('US/Eastern')
print(df_tz[['Timestamp', 'US_Eastern']])
```

```python
# Convert timezone
print("Converting to different timezones:")
df_tz['UTC'] = df_tz['US_Eastern'].dt.tz_convert('UTC')
df_tz['Europe_London'] = df_tz['US_Eastern'].dt.tz_convert('Europe/London')
df_tz['Asia_Tokyo'] = df_tz['US_Eastern'].dt.tz_convert('Asia/Tokyo')

print(df_tz[['US_Eastern', 'UTC', 'Europe_London', 'Asia_Tokyo']])
```
