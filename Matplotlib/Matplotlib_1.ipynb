```python
# Matplotlib Architecture
# Matplotlib has 3 major layers:
# Scripting Layer      -> pyplot
# Artist Layer         -> Figure, Axes, Text, Lines
# Backend Layer        -> Rendering engine
```

```python
# A. Backend Layer
# Responsible for rendering graphs:
# PNG
# PDF
# SVG
# GUI windows

# B. Artist Layer
# Everything visible on graph is an "Artist":
# Title
# Axis labels
# Lines
# Legends
# Tick labels

# C. Pyplot Layer
# Convenient command-style interface.
```

```python
# 7. Core Terminology

# Figure
# Entire canvas/window.

# Axes
# Actual plotting region.

# Axis
# X-axis / Y-axis scales.
```

```python
# 8. pyplot Interface (Stateful API)
import matplotlib.pyplot as plt

plt.plot([1,2,3],[4,5,6])
plt.title("Simple Plot")
plt.show()
```

```python
# 9. Object-Oriented Interface
fig, ax = plt.subplots()

ax.plot([1,2,3],[4,5,6])
ax.set_title("OO Style Plot")

plt.show()

# Why preferred?
# 1.Better control
# 2.Multiple subplots
# 3.Large projects
# 4.Cleaner design
```

```python
import matplotlib.pyplot as plt

# Create figure and axes
fig, (ax, bx) = plt.subplots(1,2, figsize=(10,5))

# Plot data
ax.plot([1,2,3,4], [10,20,15,25], marker="o")
bx.plot([11,12,13,14], [100,200,150,250], marker="o")

# Use fig for overall settings
fig.suptitle("Monthly Sales Report")

# Use ax for graph area settings
ax.set_title("Sales Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")

# Use ax for graph area settings
bx.set_title("Sales Trend")
bx.set_xlabel("Month")
bx.set_ylabel("Sales")

plt.show()
```

```python
# 10. First Complete Example
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,20,18,25]

plt.plot(x, y)
plt.title("Trend Analysis")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

```python
# 11. Line Customization
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,1,8,7]

plt.plot(x, y,
         color='green',
         linestyle=':',
         linewidth=3,
         marker='o',
         markersize=8)

plt.title("Customized Line Plot")
plt.show()
```

```python
# 12. NumPy Integration
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10,1)
y = x**2

plt.plot(x, y)
plt.title("Quadratic Function")
plt.show()

# Useful in:
# numerical methods
# simulations
# optimization
```

```python
# 13. Graphing Functions.
# For quadratic function:y=x*x
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = x**2

plt.plot(x,y)
plt.title("y = x²")
plt.grid(True)
plt.show()
# Parabola equation
```

```python
# Sine Wave Example
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Sine Wave")
plt.show()
```

```python
# Grid, Legend, Limits
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y,label="Sales")
plt.grid(True)
plt.legend()
plt.xlim(1,4)
plt.ylim(0,30)
plt.show()
```
