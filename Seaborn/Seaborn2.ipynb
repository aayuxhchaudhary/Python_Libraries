```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month":["Jan","Feb","Mar","Apr","May"],
    "Sales":[200,220,250,270,300]
}

df = pd.DataFrame(data)

sns.lineplot(
    data=df,
    x="Month",
    y="Sales"
)

plt.show()
```

```python
# What Happens Internally?
# Seaborn internally:
#  Reads DataFrame
#       ↓
#  Extracts x and y columns
#       ↓
#  Applies statistical estimation
#       ↓
#  Calls Matplotlib backend
#       ↓
#  Renders styled visualization
```

```python
# Understanding Confidence Interval
# This is the biggest conceptual difference between Matplotlib and Seaborn.
```

```python
# What is Confidence Interval?
# Seaborn automatically estimates uncertainty around the mean.
# Shown as:shaded region around line
```

```python
# Statistical Interpretation
# Confidence interval answers:
# “How certain are we about the estimated trend?”
```

```python
# Important Insight
# Wider confidence interval means:
# ✔ more variability
# ✔ less certainty
# Narrow interval means:
# ✔ more stable estimate
```

```python
# Disabling Confidence Interval
sns.lineplot(
    data=df,
    x="Month",
    y="Sales",
    errorbar=None
)
```

```python
# In Seaborn’s lineplot, the errorbar parameter controls whether and how uncertainty intervals 
# (like confidence intervals or standard deviations) are drawn around the line. By default, Seaborn often shows a shaded band
# around the line to represent variability.
# Explanation of errorbar
# -errorbar="ci" (default) → Plots a confidence interval (usually 95%) around the line.
# -errorbar=("ci", 68) → Plots a 68% confidence interval instead of 95%.
# -errorbar="sd" → Plots the standard deviation around the line.
# -errorbar=None → Disables error bars entirely, so you only see the line itself without any shaded uncertainty region.

# In above we used errorbar=None?
# If our data is already aggregated (e.g., monthly totals of sales), showing confidence intervals doesn’t make sense.
# It makes the plot cleaner when you just want to emphasize the trend line without statistical uncertainty.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"] * 5,
    "Sales": [100, 120, 130, 125, 140, 150,
              90, 115, 135, 120, 145, 155,
              95, 118, 132, 128, 142, 148,
              102, 119, 129, 127, 141, 152,
              98, 116, 134, 123, 143, 151]
})

# Confidence interval (default is 95%)
sns.lineplot(data=df, x="Month", y="Sales", errorbar="ci")
plt.title("Lineplot with 95% Confidence Interval")
plt.show()

# Standard deviation
sns.lineplot(data=df, x="Month", y="Sales", errorbar="sd")
plt.title("Lineplot with Standard Deviation")
plt.show()

# Custom confidence interval (68%)
sns.lineplot(data=df, x="Month", y="Sales", errorbar=("ci", 68))
plt.title("Lineplot with 68% Confidence Interval")
plt.show()

# No error bars
sns.lineplot(data=df, x="Month", y="Sales", errorbar=None)
plt.title("Lineplot with No Error Bars")
plt.show()
```

```python
# Multi-Line Visualization using Hue
# This is one of Seaborn’s strongest features.
```

```python
data = {
    "Month":["Jan","Feb","Mar","Apr","May"]*2,
    "Sales":[200,220,250,270,300,
             180,210,230,260,280],
    "Region":["North","North","North","North","North",
              "South","South","South","South","South"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Using Hue
sns.lineplot(
    data=df,
    x="Month",
    y="Sales",
    hue="Region",
    marker='o'
)
plt.show()
# The hue="Region" argument tells Seaborn to split the data into separate groups based on the values in the Region column.
# What happens with hue:
# -Seaborn looks at the Region column (North and South).
# -It draws two separate lines: one for each region.
# -Each line gets a different color automatically assigned by Seaborn’s palette.
# -The legend is also created, showing which color corresponds to which region.
```

```python
# What is Hue?
# Hue maps categorical variable → colors.
# Example:
# North → Blue
# South → Orange
```

```python
# Why Hue is Powerful?
# Without writing separate plot commands:
# ✔ Seaborn automatically groups data
# ✔ assigns colors
# ✔ creates legend
```

```python
# Interpretation
# Students should identify:
# ✔ North region consistently higher
# ✔ Both regions increasing
# ✔ Gap between trends
```

```python
# Using Style Parameter
# Style changes line appearance.
```

```python
sns.lineplot(
    data=df,
    x="Month",
    y="Sales",
    hue="Region",
    style="Region",
    markers=True
)

plt.show()
```

```python
# Why Style is Important?
# Useful for:
# ✔ grayscale printing
# ✔ colorblind users
# ✔ better differentiation
```

```python
# Using Dashes
sns.lineplot(
    data=df,
    x="Month",
    y="Sales",
    hue="Region",
    dashes=True
)
# In Seaborn’s lineplot, the dashes parameter controls whether the lines are drawn with dashed styles instead of solid lines.
# What dashes=True means
# -When we set dashes=True, Seaborn automatically assigns different dash patterns to each line (based on the hue categories).
# -This helps visually distinguish multiple lines, especially if they’re similar in color or if you want both color and line style to encode information.
# -By default, if you don’t specify dashes, Seaborn uses solid lines.
```

```python
# Size Parameter
# Maps numerical variable → line thickness.
sns.lineplot(
    data=df,
    x="Month",
    y="Sales",
    size="Sales"
)
# The size parameter controls the thickness of the lines (and sometimes the size of markers) based on a numeric variable.
# What size="Sales" does
# -Seaborn looks at the Sales column.
# -It maps the numeric values of Sales to line thickness (larger values → thicker lines).
# -This creates a visual encoding where the importance or magnitude of the data is reflected in the line width.
```

```python
# Time Series Analysis
# Line plots are ideal for:
# ✔ sequential data
# ✔ temporal trends
# ✔ forecasting preparation
```

```python
# Example: Stock Price Analysis
data = {
    "Day":[1,2,3,4,5],
    "Price":[100,105,102,110,115]
}

df = pd.DataFrame(data)

sns.lineplot(
    data=df,
    x="Day",
    y="Price",
    marker='o'
)

plt.title("Stock Price Movement")

plt.show()
```

```python
# Machine Learning Example
# Training Accuracy
```

```python
epochs = [1,2,3,4,5]

accuracy = [60,70,78,85,90]

sns.lineplot(
    x=epochs,
    y=accuracy,
    marker='o'
)

plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```
