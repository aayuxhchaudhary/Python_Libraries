```python
# Figure Size Basic Syntax
plt.figure(figsize=(width, height))
# Units → inches
# Default size ≈ (6.4, 4.8)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot([1,2,3],[2,4,6])
plt.title("Custom Figure Size")

plt.show()
```

```python
# DPI (Dots Per Inch)
# -DPI controls resolution of the figure
plt.figure(figsize=(8,5), dpi=100)
```

```python
plt.figure(figsize=(6,4), dpi=200)
plt.plot([1,2,3],[3,6,9])
plt.title("High Resolution Plot")
plt.show()
```

```python
# | DPI     | Usage           |
# | ------- | --------------- |
# | 72      | screen display  |
# | 100–150 | presentations   |
# | 300     | research papers |
```

```python
plt.plot([1,2,3],[2,5,3], color='#FF5733')
```

```python
# Line Styling
plt.plot(x, y,
         linestyle='--',
         linewidth=2,
         marker='o')
```

```python
# Titles and Labels Styling
plt.title("Sales Trend", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
```

```python
# Font Customization
plt.rcParams['font.family'] = 'serif'
```

```python
# Grid Styling
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
# Legend Styling
plt.legend(loc='upper left', fontsize=10)
```

```python
# Axis Customization
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
# Tick Customization
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

```python
# Background Styling
plt.style.use('dark_background')
```

```python
plt.style.use('ggplot')

plt.plot([1,2,3],[2,4,6])
plt.title("GGPlot Style")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.figure(figsize=(8,5), dpi=150)

plt.plot(x, y,
         color='blue',
         linestyle='-',
         marker='o',
         linewidth=2,
         label='Sales')

plt.title("Sales Trend", fontsize=16, fontweight='bold')
plt.xlabel("Time", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
