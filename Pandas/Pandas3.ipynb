```python
# Pandas DataFrames
# What is a DataFrame?
# A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.
```

```python
# Example: Create a simple Pandas DataFrame

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

```python
# Locate Row:As you can see from the result above, the DataFrame is like a table with rows and columns.

# Pandas use the loc attribute to return one or more specified row(s)
```

```python
# Return row 0:

#refer to the row index:
print(df.loc[2])
```

```python
# Return row 0:

#refer to the row index:
print(df.loc[0])
```

```python
# Named Indexes:With the index argument, you can name your own indexes.

# Example:Add a list of names to give each row a name:

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["1", "2", "3"])

print(df) 
```

```python
# Named Indexes:With the index argument, you can name your own indexes.

# Example:Add a list of names to give each row a name:

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["101", "102", "103"])

print(df) 
```

```python
# Locate Named Indexes:Use the named index in the loc attribute to return the specified row(s).

# Example:Return "day2":

#refer to the named index:
print(df.loc['101'])
```
