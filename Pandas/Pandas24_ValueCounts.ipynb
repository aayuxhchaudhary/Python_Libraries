```python
# PART 2: UNDERSTANDING unique() METHOD
```

```python
# 2.1 Basic unique() Usage
print("="*70)
print("unique() - FINDING UNIQUE VALUES")
print("="*70)

# Create dataset
df = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Marketing', 'Finance'],
    'Status': ['Active', 'Active', 'Inactive', 'Active', 'Active', 'Inactive', 'Active', 'Active']
})

print("Dataset:")
print(df)
```

```python
# 1. Basic unique() - returns array of unique values
print("\n Unique departments:")
unique_depts = df['Department'].unique()
print(unique_depts)
print(f"Type: {type(unique_depts)}")
print(f"Number of unique departments: {len(unique_depts)}")
```

```python
# 2. unique() preserves order of appearance
print("Order of appearance:")
print(f"Unique values appear in order: {unique_depts}")
```

```python
# 3. Using unique() with different data types
print("Unique statuses:")
print(df['Status'].unique())
```

```python
# 4. unique() with numbers
numbers = pd.Series([10, 20, 10, 30, 20, 40, 10, 30])
print(f"\nOriginal numbers: {numbers.tolist()}")
print(f"Unique numbers: {numbers.unique()}")
```

```python
# 5. Handling missing values
with_nan = pd.Series(['A', 'B', 'A', np.nan, 'B', np.nan, 'C'])
print(f"\nSeries with NaN: {with_nan.tolist()}")
print(f"Unique (includes NaN): {with_nan.unique()}")
```

```python
# 2.2 Practical Applications of unique()
print("="*70)
print("PRACTICAL APPLICATIONS OF unique()")
print("="*70)

# Create product data
products = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104, 105, 106],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Clothing', 'Electronics'],
    'Size': ['M', 'L', 'XL', None, 'M', 'L']
})

print("Product Data:")
print(products)
```

```python
# 1. Data validation - check for unexpected values
print("Data Validation:")
expected_categories = ['Electronics', 'Clothing', 'Books']
actual_categories = products['Category'].unique()
print(f"Expected: {expected_categories}")
print(f"Actual: {actual_categories}")

unexpected = set(actual_categories) - set(expected_categories)
if unexpected:
    print(f"Unexpected categories found: {unexpected}")
else:
    print(" All categories are valid")
```

```python
# 2. Finding unique combinations
print("Unique Category-Size combinations:")
unique_combos = products[['Category', 'Size']].drop_duplicates()
print(unique_combos)
```

```python
# 3. Creating dropdown/selection lists
print("Creating selection lists for UI:")
categories = products['Category'].unique()
sizes = products['Size'].dropna().unique()
print(f"Category dropdown: {categories.tolist()}")
print(f"Size dropdown: {sizes.tolist()}")
```

```python
# 4. Data quality check - find missing categories
print("Data Quality Check:")
if products['Category'].isnull().any():
    print("Warning: Missing categories found!")
else:
    print("No missing categories")
```

```python
# 5. Converting unique values to list for further processing
print("Converting to list for reporting:")
category_list = products['Category'].unique().tolist()
print(f"Categories as list: {category_list}")
print(f"Number of categories: {len(category_list)}")
```

```python
# PART 3: UNDERSTANDING nunique() METHOD
```

```python
# 3.1 Basic nunique() Usage
print("="*70)
print("nunique() - COUNTING UNIQUE VALUES")
print("="*70)

# Create dataset
sales = pd.DataFrame({
    'Store': ['A', 'B', 'A', 'C', 'B', 'A', 'D', 'C'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Monitor', 'Tablet', 'Keyboard'],
    'Customer': [101, 102, 101, 103, 102, 104, 105, 103]
})

print("Sales Data:")
print(sales)
```

```python
# 1. Basic nunique() - count unique values in a column
print("Count unique values:")
print(f"Unique stores: {sales['Store'].nunique()}")
print(f"Unique products: {sales['Product'].nunique()}")
print(f"Unique customers: {sales['Customer'].nunique()}")
```

```python
# 2. nunique() on entire DataFrame
print("Unique counts for all columns:")
print(sales.nunique())
```

```python
# 3. nunique() with dropna parameter
print("Handling missing values:")
with_nulls = pd.Series(['A', 'B', 'A', None, 'B', None, 'C'])
print(f"Series: {with_nulls.tolist()}")
print(f"Unique count (include NaN): {with_nulls.nunique(dropna=False)}")
print(f"Unique count (exclude NaN): {with_nulls.nunique(dropna=True)}")
```

```python
# 4. nunique() on specific axis
print("nunique() on rows (axis=1):")
print(f"Unique values per row:\n{sales.nunique(axis=1)}")
```

```python
# 5. Percentage of unique values
print("Percentage analysis:")
total_rows = len(sales)
for col in sales.columns:
    unique_pct = (sales[col].nunique() / total_rows) * 100
    print(f"{col}: {unique_pct:.1f}% unique")
```

```python
# 4.1 Basic value_counts() Usage
print("="*70)
print("value_counts() - FREQUENCY DISTRIBUTION")
print("="*70)

# Create survey data
survey = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 
                'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop'],
    'Rating': [5, 4, 5, 3, 4, 5, 5, 4, 3, 5],
    'City': ['NY', 'LA', 'NY', 'CHI', 'LA', 'NY', 'HOU', 'LA', 'CHI', 'NY']
})

print("Survey Data:")
print(survey)
```

```python
# 1. Basic value_counts()
print("Product frequency:")
product_counts = survey['Product'].value_counts()
print(product_counts)
```

```python
# 2. value_counts() with percentages
print("Product frequency (percentages):")
product_percent = survey['Product'].value_counts(normalize=True) * 100
print(product_percent.round(2))
```

```python
# 3. value_counts() sorted alphabetically
print("Product frequency (alphabetical order):")
product_alpha = survey['Product'].value_counts().sort_index()
print(product_alpha)
```

```python
# 4. value_counts() in ascending order
print("Product frequency (ascending):")
product_asc = survey['Product'].value_counts(ascending=True)
print(product_asc)
```

```python
# 5. Including only top 2 values
print("Top 2 most frequent products:")
top_products = survey['Product'].value_counts().head(2)
print(top_products)
```

```python
# 6. Excluding bottom values
print("Bottom 2 least frequent products:")
bottom_products = survey['Product'].value_counts().tail(2)
print(bottom_products)
```

```python
# 4.2 Handling Missing Values with value_counts()
print("="*70)
print("value_counts() WITH MISSING VALUES")
print("="*70)

# Create data with missing values
data_with_na = pd.Series(['A', 'B', 'A', np.nan, 'B', 'A', np.nan, 'C', 'B', np.nan])

print(f"Data: {data_with_na.tolist()}")
```

```python
# 1. Default - excludes NaN
print("Default (excludes NaN):")
print(data_with_na.value_counts())
```

```python
# 2. Include NaN in counts
print("Include NaN (dropna=False):")
print(data_with_na.value_counts(dropna=False))
```

```python
# 3. Percentage including NaN
print("Percentages including NaN:")
print(data_with_na.value_counts(normalize=True, dropna=False) * 100)
```

```python
# 4. Compare with and without NaN
print("Impact of NaN on analysis:")
print(f"Without NaN: {data_with_na.value_counts().sum()} values")
print(f"With NaN: {data_with_na.value_counts(dropna=False).sum()} values")
```

```python
# 4.3 Binning with value_counts()
print("="*70)
print("BINNING WITH value_counts()")
print("="*70)

# Create age data
ages = pd.Series([25, 32, 45, 28, 35, 42, 38, 29, 51, 33, 27, 44, 39, 31, 48, 26, 41, 36, 52, 29])

print(f"Ages: {ages.tolist()}")
```

```python
# 1. Basic binning with value_counts()
print("Age groups (bins of 10):")
age_groups = pd.cut(ages, bins=[20, 30, 40, 50, 60])
print(age_groups.value_counts())
```

```python
# 2. Custom bins with labels
print("Age groups with custom labels:")
age_labels = pd.cut(ages, bins=[20, 30, 40, 50, 60], 
                    labels=['20s', '30s', '40s', '50s'])
print(age_labels.value_counts())
```

```python
# 3. Percentages for bins
print("Age distribution percentages:")
age_pct = age_labels.value_counts(normalize=True) * 100
print(age_pct.round(1))
```

```python
# 4. Creating bins with equal number of items (qcut)
print("Quartiles (equal sized groups):")
age_quartiles = pd.qcut(ages, q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
print(age_quartiles.value_counts())
```
