```python
# Advanced Styling
# Example Dataset
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]
sales = [200,220,250,270,300]
```

```python
# Basic Styled Plot
plt.figure(figsize=(8,5), dpi=150)

plt.plot(
    months,
    sales,
    color='blue',
    linestyle='--',
    linewidth=3,
    marker='o',
    markersize=8
)

plt.title("Monthly Sales Trend",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.grid(True, alpha=0.3)

plt.show()
```

```python
# Explanation of Styling Parameters
# | Parameter | Purpose         |
# | --------- | --------------- |
# | color     | line color      |
# | linestyle | visual pattern  |
# | linewidth | thickness       |
# | marker    | point indicator |
# | fontsize  | text size       |
# | alpha     | transparency    |
```

```python
# Annotations (Very Important)
# -Annotations highlight important events or points.
```

```python
# Example: Annotating Peak Sales
plt.plot(months, sales, marker='o')

plt.annotate(
    "Highest Sales",
    xy=("May",300),
    xytext=("Mar",280),
    arrowprops=dict(facecolor='black')
)

plt.show()
```

```python
# Why Annotations are Important?
# Used in:
# -research papers
# -dashboards
# -reports
# -business presentations
# They guide audience attention.
```

```python
# # Annotation Parameters
# | Parameter  | Meaning             |
# | ---------- | ------------------- |
# | xy         | target point        |
# | xytext     | annotation position |
# | arrowprops | arrow styling       |
```

```python
# Subplots Dashboards
# -Dashboards combine multiple visualizations.
```

```python
# Why Dashboards?
# Real-world analytics requires:
# -multiple perspectives
# -summary visualization
# -compact reporting
```

```python
# Example Dashboard
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month":["Jan","Feb","Mar","Apr","May"],
    "Sales":[200,220,250,270,300],
    "Profit":[50,60,70,65,80]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(2,2, figsize=(10,8))

# Line Plot
ax[0,0].plot(df["Month"], df["Sales"])
ax[0,0].set_title("Sales Trend")

# Bar Plot
ax[0,1].bar(df["Month"], df["Profit"])
ax[0,1].set_title("Profit")

# Histogram
ax[1,0].hist(df["Sales"])
ax[1,0].set_title("Sales Distribution")

# Scatter Plot
ax[1,1].scatter(df["Sales"], df["Profit"])
ax[1,1].set_title("Sales vs Profit")

plt.tight_layout()
plt.show()
```

```python
# Interpretation of Dashboard
# -sales trend increasing
# -profit variability
# -distribution shape
# -correlation between sales and profit
```
