```python
# GroupBy & Aggregation in pandas
import pandas as pd

df = pd.read_json("data1.json")
print(df)
```

```python
# What is GroupBy?
# GroupBy means grouping rows based on a column and applying operations.
```

```python
df.groupby("city")

# This does NOT show output directly.It creates a grouped object.
```

```python
# Mean marks by city
# vc=print(df.groupby("city")["marks"].mean())
# print(vc)
(df.groupby("city")["marks"].mean()) as MeanMarks
print()
# bc=print(df.groupby("city")["age"].mean()) as MeanAge
# print(bc)
(df.groupby("city")["age"].mean()) AS MeanAge
```

```python
# Mean marks by city
print(df.groupby("city")["marks"].mean().reset_index(name="MeanMarks"))
# print(vc)

print(df.groupby("city")["age"].mean().reset_index(name="MeanAge"))
# print(bc)
```

```python
# Mean marks by city
print(df.groupby("city")["marks"].mean())
```

```python
# Multiple Aggregations
import pandas as pd

df = pd.read_json("data1.json")

print(df.groupby("age")["marks"].agg(["mean","median", "max", "min"]))
```

```python
# Multiple Aggregations
import pandas as pd

df = pd.read_json("data1.json")

print(df.groupby("city")["marks"].agg(["mean", "max", "min"]))
```

```python
# Count Values
import pandas as pd

df = pd.read_json("data1.json")

print(df.groupby("city")["name"].count())

# Count students per city.
```

```python
# Group by Multiple Columns
print(df.groupby(["city", "marks"])["name"].count())

# Hierarchical grouping.
```

```python
# Reset Index
result = df.groupby("city")["marks"].mean().reset_index()
print(result)

# Converts grouped result back to DataFrame.
```

```python
# Sorting Group Results
print(df.groupby("city")["marks"].mean().sort_values(ascending=False))
```

```python
# Named Aggregation

df.groupby("city").agg(
    avg_marks=("marks", "mean"),
    max_marks=("marks", "max")
)
```

```python
# Apply Custom Function
df.groupby("city")["marks"].apply(lambda x: x.max() - x.min())

# Difference between highest and lowest marks.
```

```python
# Size vs Count
df.groupby("city").size()
```

```python
df.groupby("city")["name"].count()
```

```python
df.groupby("city").count()
```

```python
# Difference
# size → counts all rows
# count → ignores null values
```
