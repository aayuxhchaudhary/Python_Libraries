```python
# Handling Missing Values in Pandas
# Missing values = data not available
# Represented as:
# NaN (Not a Number)
# None
```

```python
import pandas as pd

df = pd.read_json("student2.json")

print(df)
```

```python
# Detect Missing Values
# isnull()
import pandas as pd

df = pd.read_json("student2.json")

print(df.isnull())
```

```python
# True → missing
# False → present
```

```python
# Count missing values

import pandas as pd

df = pd.read_json("student2.json")

print(df.isnull().sum())
```

```python
# notnull():Opposite of isnull.
import pandas as pd

df = pd.read_json("student2.json")

print(df.notnull())
```

```python
# Drop Missing Values
# dropna():Remove rows with missing values.
import pandas as pd

df = pd.read_json("student2.json")

df1 = df.dropna()

print(df1)

# Removes any row containing at least one null.
```

```python
# Drop only specific columns
import pandas as pd

df = pd.read_json("student2.json")

df.dropna(subset=["marks","age"])

# Only remove rows where marks is missing.
```

```python
# axis parameter
import pandas as pd

df = pd.read_json("student2.json")

sb = df.dropna(axis=1)

print(sb)
# Removes columns with null values.
```

```python
# axis parameter
import pandas as pd

df = pd.read_json("student2.json")

sb = df.dropna(axis=0)

print(sb)
# Removes columns with null values.
```

```python
# how parameter:Removes rows only if ALL values are null.
import pandas as pd

df = pd.read_json("student2.json")

df.dropna(how="all")
```

```python
# how parameter:Removes rows only if ALL values are null.
import pandas as pd

df = pd.read_json("student2.json")

vc = df.dropna(how="all")

print(vc)
```

```python
# Fill Missing Values
# fillna():Replace missing values.
import pandas as pd

df = pd.read_json("student2.json")

df["age"].fillna(24)

# print(df)
```

```python
# Fill Missing Values
# fillna():Replace missing values.
import pandas as pd

df = pd.read_json("student2.json")

df["age"].fillna(24)

# print(df)
```

```python
import pandas as pd

df = pd.read_json("student2.json")

df[["age", "marks"]] = df[["age", "marks"]].fillna({
    "age": 24.0,
    "marks": 33.0
})
print(df)
```

```python
import pandas as pd

df = pd.read_json("student2.json")

df["marks"].fillna(df["marks"].mean(), inplace=True)

print(df)

# Mean = average of column.
```

```python
import pandas as pd

df = pd.read_json("student2.json")

df["marks"].fillna(df["marks"].median(), inplace=True)

print(df)
```

```python
# Fill with mode:Mode = most frequent value.#takes value in ascending order

import pandas as pd

df = pd.read_json("student2.json")

df["marks"].fillna(df["marks"].mode()[1], inplace=True)

print(df)
```

```python
# Fill with mode:Mode = most frequent value.#takes value in ascending order

import pandas as pd

df = pd.read_json("student2.json")

df["city"].fillna(df["city"].mode()[0], inplace=True)

print(df)
```

```python
# Forward Fill (ffill):Takes previous value.
import pandas as pd

df = pd.read_json("student2.json")

df.fillna(method="ffill", inplace=True)

print(df)
```

```python
# Backward Fill (bfill):Takes next value.
import pandas as pd

df = pd.read_json("student2.json")

df.fillna(method="bfill", inplace=True)

print(df)
```

```python
# Replace Missing Values in Entire DataFrame
import pandas as pd

df = pd.read_json("student2.json")

df.fillna(0, inplace=True)

print(df)
```

```python
#Check Again After Filling
import pandas as pd

df = pd.read_json("student2.json")

print(df.isnull().sum())

print(df)
```
