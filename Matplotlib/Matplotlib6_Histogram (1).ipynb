```python
# 2. What is a Histogram?

# A Histogram is a graphical representation of the distribution of numerical data.
# It groups data into intervals called bins and shows:
```

```python
# Key Concepts
# -1. Bins
# -Intervals dividing data:
# -Example:0–10, 10–20, 20–30

# 2. Frequency
# -Number of data points in each bin.

# 3. Range
# -Difference between max and min values.
```

```python
import matplotlib.pyplot as plt

data = [45,50,55,60,62,65,68,70,72,75,80,85]

plt.hist(data, edgecolor='black')
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [45,50,55,60,62,65,68,70,72,75,80,85]

plt.hist(marks, bins=5, color='blue', edgecolor='black')

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
# Interpretation
# -Most students between 60–75
# -Few high scorers
# -Distribution spread
```

```python
# Effect of Bin Size
plt.hist(marks, bins=3, edgecolor='black')
plt.show()

plt.hist(marks, bins=10, edgecolor='black')
plt.show()
```

```python
# Types of Distributions
# A. Normal Distribution
# -Symmetrical
# -Bell-shaped
```

```python
# Histogram with Curve
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,1000)

plt.hist(data, bins=30, edgecolor='black' ,density=True)

x = np.linspace(-4,4,100) #Creates 100 evenly spaced points between -4 and 4.
y = (1/(np.sqrt(2*np.pi))) * np.exp(-0.5*x**2)

plt.plot(x,y)

plt.show()
```

```python
# B. Right Skewed (Positive Skew)
# -Tail on right
# -Majority values on left
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate right-skewed data
data = np.random.exponential(scale=1, size=1000)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')

# Labels and title
plt.title("Right Skewed (Positive Skew) Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# C. Left Skewed (Negative Skew)
# -Tail on left
# -Majority values on right
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate right-skewed data
data = np.random.exponential(scale=1, size=1000)

# Step 2: Convert to left-skewed
data = -data   # flip distribution

# Step 3: Plot histogram
plt.hist(data, bins=30, edgecolor='black')

# Labels and title
plt.title("Left Skewed (Negative Skew) Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# Explanation:
# 1. Exponential Distribution
# Produces right-skewed data
# Most values near 0, few large values
# 2. Negation (data = -data)
# Flips the data across zero
# Right tail becomes left tail
# This creates negative skew
```

```python
# Using NumPy for Random Data
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution")
plt.show()

# random.normal(0,1,1000):1,000 samples from a normal distribution with mean = 0 and standard deviation = 1.
# The result is a dataset that should look like a bell curve when plotted.
```

```python
# Density Histogram
plt.hist(data, bins=30, edgecolor='black', density=True)

# Converts frequency to probability
# Area = 1
```

```python
# What density=True Does
# -Normally, plt.hist shows counts: the number of data points falling into each bin.
# -With density=True, the histogram is normalized so that the total area under all bars equals 1.
# -Each bar’s height now represents a probability density, not raw counts.
```

```python
# Outlier Detection
data = [10,12,13,15,14,16,18,100]

plt.hist(data, bins=5, edgecolor='black')
plt.show()

# Interpretation
# -100 is outlier
# -isolated bin
```

```python
# Multiple Histograms
data1 = np.random.normal(0,1,1000)
data2 = np.random.normal(2,1,1000)

plt.hist(data1, bins=30, alpha=0.5,edgecolor='black', label='Data1')
plt.hist(data2, bins=30, alpha=0.5,edgecolor='blue', label='Data2')

plt.legend()
plt.show()
```

```python
# Interpretation:
# -Compare distributions
# -Shift in mean
# -Spread differences
```
