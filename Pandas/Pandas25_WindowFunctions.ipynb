```python
# PART 1: INTRODUCTION TO WINDOW FUNCTIONS
# What are Window Functions?
# Window functions perform calculations across a set of rows that are related to the current row. 
# Unlike regular functions, they don't collapse rows - they maintain the original number of rows.
```

```python
import pandas as pd
import numpy as np

print("="*70)
print("WHAT ARE WINDOW FUNCTIONS?")
print("="*70)

# Create sample sales data
sales = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=10, freq='D'),
    'Daily_Sales': [100, 120, 110, 130, 125, 140, 135, 150, 145, 160]
})

print("Daily Sales Data:")
print(sales)
```

```python
# WITHOUT window function - regular calculation
print("WITHOUT WINDOW FUNCTION (collapses data):")
average_sales = sales['Daily_Sales'].mean()
print(f"Average Sales: {average_sales}")
print("Result: Only 1 value (lost all daily detail)")
```

```python
# WITH window function - maintains rows
print("\n WITH WINDOW FUNCTION (maintains rows):")
sales['7_Day_Avg'] = sales['Daily_Sales'].rolling(window=7).mean()
print(sales)
print("\nResult: Still 10 rows, each with its own 7-day average")

print("\n KEY CONCEPT:")
print("Window functions calculate values based on a 'window' of rows")
print("- Rolling: Fixed size window that moves")
print("- Expanding: Growing window from start")
print("- Exponential: Weighted window (more weight to recent)")
```

```python
# Types of Window Functions
```

```python
print("="*70)
print("THREE TYPES OF WINDOW FUNCTIONS")
print("="*70)

# Create sample data
data = pd.DataFrame({
    'Day': range(1, 11),
    'Value': [10, 20, 15, 25, 30, 28, 35, 32, 40, 38]
})

print("Original Data:")
print(data)
```

```python
# 1. Rolling Window (fixed size, moves)
print("1️. ROLLING WINDOW (window=3):")
data['Rolling_Mean'] = data['Value'].rolling(window=3).mean()
print(data[['Day', 'Value', 'Rolling_Mean']])
```

```python
# 2. Expanding Window (grows from start)
print("2. EXPANDING WINDOW:")
data['Expanding_Mean'] = data['Value'].expanding().mean()
print(data[['Day', 'Value', 'Expanding_Mean']])
```

```python
# 3. Exponential Weighted Window
print("3️. EXPONENTIAL WEIGHTED WINDOW (span=3):")
data['EWM_Mean'] = data['Value'].ewm(span=3,adjust=False).mean()
print(data[['Day', 'Value', 'EWM_Mean']])


# a=2/span + 1 = 2/4 = 0.5
# EWMt = a.valuet + (1-a).EWMt-1
print("\n COMPARISON:")
print("• Rolling: Last 3 days average (equal weight)")
print("• Expanding: Average of ALL days so far")
print("• EWM: Weighted average (recent days count more)")
```

```python
# PART 2: ROLLING WINDOW FUNCTIONS
```

```python
# 2.1 Basic Rolling Window Operations
print("="*70)
print("ROLLING WINDOW FUNCTIONS")
print("="*70)

# Create daily temperature data
temperatures = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=15, freq='D'),
    'Temp': [22, 23, 21, 24, 25, 23, 22, 26, 27, 25, 24, 26, 28, 27, 29]
})
temperatures.set_index('Date', inplace=True)

print("Daily Temperatures:")
print(temperatures)
```

```python
# 1. Rolling Mean (Moving Average)
print("3-Day Moving Average:")
temperatures['MA_3'] = temperatures['Temp'].rolling(window=3).mean()
print(temperatures[['Temp', 'MA_3']].head(10))
# .head(10):“Show me the first 10 rows of the DataFrame”
```

```python
# 2. Rolling Sum
print("7-Day Rolling Sum:")
temperatures['Sum_7'] = temperatures['Temp'].rolling(window=7).sum()
print(temperatures[['Temp', 'Sum_7']].head(12))
```

```python
# 3. Rolling Maximum
print("5-Day Rolling Maximum:")
temperatures['Max_5'] = temperatures['Temp'].rolling(window=5).max()
print(temperatures[['Temp', 'Max_5']].head(10))
```

```python
# 4. Rolling Minimum
print("5-Day Rolling Minimum:")
temperatures['Min_5'] = temperatures['Temp'].rolling(window=5).min()
print(temperatures[['Temp', 'Min_5']].head(10))
```

```python
# 5. Rolling Standard Deviation (Volatility)
print("7-Day Rolling Standard Deviation:")
temperatures['Std_7'] = temperatures['Temp'].rolling(window=7).std()
print(temperatures[['Temp', 'Std_7']].head(10))
```

```python
# 6. Rolling Count (non-null values)
print("Rolling Count:")
temperatures['Count_5'] = temperatures['Temp'].rolling(window=5).count()
print(temperatures[['Temp', 'Count_5']].head(10))
```

```python
# 2.2 Rolling Window with min_periods
```

```python
print("="*70)
print("HANDLING STARTING PERIODS (min_periods)")
print("="*70)

# Create data
data = pd.DataFrame({
    'Day': range(1, 11),
    'Sales': [100, 120, 110, 130, 125, 140, 135, 150, 145, 160]
})

print("Original Data:")
print(data)
```

```python
# Default: NaN for first window-1 rows
print("Default (min_periods=None):")
data['MA_Default'] = data['Sales'].rolling(window=5).mean()
print(data[['Sales', 'MA_Default']])
```

```python
# With min_periods=2 (start calculating with only 2 values)
print("With min_periods=2:")
data['MA_Min2'] = data['Sales'].rolling(window=5, min_periods=2).mean()
print(data[['Sales', 'MA_Min2']])
```

```python
# With min_periods=2 (start calculating with only 2 values)
print("With min_periods=2:")
data['MA_Min2'] = data['Sales'].rolling(window=5).mean()
print(data[['Sales', 'MA_Min2']])
```

```python
# With min_periods=1 (start calculating from first row)
print("With min_periods=1:")
data['MA_Min1'] = data['Sales'].rolling(window=5, min_periods=1).mean()
print(data[['Sales', 'MA_Min1']])

print("\n WHEN TO USE min_periods:")
print("• min_periods=1: Get values from day 1 (less accurate initially)")
print("• min_periods=window: Only get values after full window (more accurate)")
print("• Lower min_periods = fewer NaN values but less reliable early results")
```

```python
# 2.3 Rolling Window on Multiple Columns
print("="*70)
print("ROLLING WINDOW ON MULTIPLE COLUMNS")
print("="*70)

# Create multi-column data
stock_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=10, freq='D'),
    'AAPL': [150, 152, 151, 153, 155, 154, 156, 158, 157, 159],
    'GOOGL': [2800, 2820, 2810, 2830, 2850, 2840, 2860, 2880, 2870, 2890],
    'MSFT': [300, 305, 302, 308, 310, 307, 312, 315, 313, 318]
})
stock_data.set_index('Date', inplace=True)

print("Stock Prices:")
print(stock_data)
```

```python
# 1. Rolling mean on all columns
print("3-Day Moving Average for all stocks:")
ma_3 = stock_data.rolling(window=3).mean()
print(ma_3.round(2))
```

```python
# 2. Rolling standard deviation
print("3-Day Rolling Volatility:")
volatility = stock_data.rolling(window=3).std()
print(volatility.round(2))
```

```python
# 3. Different windows for different columns
print("Different windows per column:")
result = pd.DataFrame({
    'AAPL_MA5': stock_data['AAPL'].rolling(5).mean(),
    'GOOGL_MA3': stock_data['GOOGL'].rolling(3).mean(),
    'MSFT_MA7': stock_data['MSFT'].rolling(7).mean()
})
print(result.round(2))
```

```python
# 4. Rolling correlation between stocks
print("5-Day Rolling Correlation (AAPL vs GOOGL):")
correlation = stock_data['AAPL'].rolling(5).corr(stock_data['GOOGL'])
print(correlation.round(4))#explained below
```

```python
# 2.4 Rolling Window with Center Parameter
print("="*70)
print("CENTERED ROLLING WINDOWS")
print("="*70)

# Create data
data = pd.DataFrame({
    'Day': range(1, 11),
    'Value': [10, 12, 11, 13, 14, 12, 15, 16, 14, 17]
})

print("Original Data:")
print(data)
```

```python
# Default: window ends at current row
print("Default (center=False):")
data['MA_Default'] = data['Value'].rolling(window=3).mean()
print(data[['Value', 'MA_Default']])
```

```python
# Centered: window centered on current row
print("Centered (center=True):")
data['MA_Center'] = data['Value'].rolling(window=3, center=True).mean()
print(data[['Value', 'MA_Center']])

print("\n VISUAL EXPLANATION:")
print("Day 3 (Default): Average of Days 1,2,3")
print("Day 3 (Centered): Average of Days 2,3,4")
print("\n Use center=True for smoothing without lag")
print("Use center=False for forecasting (no future data)")
```

```python
# PART 3: EXPANDING WINDOW FUNCTIONS
```

```python
# 3.1 Basic Expanding Window Operations
print("="*70)
print("EXPANDING WINDOW FUNCTIONS (Cumulative)")
print("="*70)

# Create sales data
sales = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [1000, 1200, 1100, 1300, 1250, 1400]
})

print("Monthly Sales:")
print(sales)
```

```python
# 1. Expanding Sum (Cumulative Sum)::Take all data from the beginning up to current row and apply function”
print("\n Cumulative Sales (Year-to-Date):")
sales['Cumulative_Sales'] = sales['Sales'].expanding().sum()
print(sales)
```

```python
# 2. Expanding Mean (Running Average)
print("\n Running Average Sales:")
sales['Running_Avg'] = sales['Sales'].expanding().mean()
print(sales[['Sales', 'Running_Avg']])
```

```python
# 3. Expanding Maximum
print("\n Running Maximum:")
sales['Running_Max'] = sales['Sales'].expanding().max()
print(sales[['Sales', 'Running_Max']])
```

```python
# 4. Expanding Minimum
print("\n Running Minimum:")
sales['Running_Min'] = sales['Sales'].expanding().min()
print(sales[['Sales', 'Running_Min']])
```

```python
# 5. Expanding Standard Deviation
print("\n Running Standard Deviation:")
sales['Running_Std'] = sales['Sales'].expanding().std()
print(sales[['Sales', 'Running_Std']].round(2))
```

```python
# 3.2 Expanding Window Applications
print("="*70)
print("EXPANDING WINDOW APPLICATIONS")
print("="*70)

# Create investment data
investment = pd.DataFrame({
    'Year': range(2015, 2025),
    'Annual_Return': [1000, 1200, 1150, 1300, 1250, 1400, 1350, 1500, 1450, 1600]
})

print("Annual Investment Returns:")
print(investment)
```

```python
# 1. Running Total (Portfolio Value)
print("Portfolio Value Over Time:")
investment['Portfolio_Value'] = investment['Annual_Return'].expanding().sum()
print(investment[['Year', 'Annual_Return', 'Portfolio_Value']])
```

```python
# 2. Average Return (Performance Trend)
print("Average Annual Return:")
investment['Avg_Return'] = investment['Annual_Return'].expanding().mean()
print(investment[['Year', 'Annual_Return', 'Avg_Return']])
```

```python
# 3. Best Year So Far
print("Best Performance So Far:")
investment['Best_Year'] = investment['Annual_Return'].expanding().max()
investment['Is_Record'] = investment['Annual_Return'] == investment['Best_Year']
print(investment[['Year', 'Annual_Return', 'Best_Year', 'Is_Record']])
```

```python
# 4. Growth from Start
print("Growth from Initial Investment:")
initial = investment['Annual_Return'].iloc[0]
investment['Growth_Pct'] = (investment['Portfolio_Value'] / initial - 1) * 100
print(investment[['Year', 'Portfolio_Value', 'Growth_Pct']].round(1))
```

```python
# 5. Rolling vs Expanding Comparison
print("Rolling (3-year) vs Expanding (all years):")
investment['Rolling_3_Avg'] = investment['Annual_Return'].rolling(3).mean()
investment['Expanding_Avg'] = investment['Annual_Return'].expanding().mean()
print(investment[['Year', 'Annual_Return', 'Rolling_3_Avg', 'Expanding_Avg']].round(2))
```

```python
# 4.1 Understanding Exponential Weighting
print("="*70)
print("EXPONENTIAL WEIGHTED FUNCTIONS (EWM)")
print("="*70)

# Create data with recent trend
data = pd.DataFrame({
    'Day': range(1, 21),
    'Value': [10, 11, 10, 12, 11, 13, 12, 14, 13, 15,
              14, 16, 15, 17, 16, 18, 17, 19, 18, 50]  # Sudden spike at end
})

print("Data with Recent Spike:")
print(data.tail(10))
```

```python
# 1. Simple Moving Average (equal weight)
print("Simple Moving Average (window=5):")
data['SMA_5'] = data['Value'].rolling(5).mean()
print(data[['Value', 'SMA_5']].tail(10))
```

```python
# 2. Exponential Weighted Moving Average (more weight to recent)
print("Exponential Weighted MA (span=5):")
data['EWMA_5'] = data['Value'].ewm(span=5).mean()
print(data[['Value', 'SMA_5', 'EWMA_5']].tail(10))

print("\n WHY EWM IS DIFFERENT:")
print("• SMA: All 5 days have EQUAL weight (20% each)")
print("• EWM: Recent days have MORE weight")
print("• EWM reacts faster to recent changes (like the spike to 50)")
```

```python
# 3. Comparing different spans
print("Different Span Values (recent weight):")
data['EWMA_3'] = data['Value'].ewm(span=3).mean()   # More reactive
data['EWMA_10'] = data['Value'].ewm(span=10).mean() # Less reactive
print(data[['Value', 'EWMA_3', 'EWMA_10']].tail(10))
```
