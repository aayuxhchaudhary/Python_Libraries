```python
# df.to_csv() - Export to CSV Format
# What is CSV?
# CSV (Comma-Separated Values) is a simple text format where data is separated by commas. It's like a digital table stored as plain text.
```

```python
import pandas as pd
import numpy as np

# Create a sample dataset of students
students_data = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice Johnson', 'Bob Smith', 'Carol Davis', 'David Wilson', 'Emma Brown'],
    'Age': [20, 21, 19, 22, 20],
    'Grade': ['A', 'B', 'A+', 'C', 'B+'],
    'Marks': [85.5, 72.3, 92.0, 65.5, 78.8],
    'Attendance': [95, 88, 98, 75, 90]
}

df = pd.DataFrame(students_data)
print("Original DataFrame:")
print(df)
# print("\n" + "="*50 + "\n")
```

```python
# Basic CSV Export
# python
# Simple export with default settings
df.to_csv('students_basic.csv')
print("Created: students_basic.csv")
```
