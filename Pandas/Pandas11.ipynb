```python
# Merging & Joining DataFrames in pandas
```

```python
import pandas as pd

df_students = pd.read_json("student3.json")
df_marks = pd.read_json("marks.json")

print(df_students)
print(df_marks)
```

```python
# What is Merge?
# Merge = combining two DataFrames based on a common column (like SQL JOIN)
```

```python
# Inner Join 

df = pd.merge(df_students, df_marks, on="id", how="inner")

print(df)
# Only matching IDs are included.
```

```python
# Left Join
df = pd.merge(df_students, df_marks, on="id", how="left")

print(df)
# All students kept, missing marks = NaN
```

```python
# Right Join
df = pd.merge(df_students, df_marks, on="id", how="right")

print(df)
# All marks kept, even if student missing.
```

```python
# Outer Join

df = pd.merge(df_students, df_marks, on="id", how="outer")

print(df)

# All data from both tables.
```

```python
# Summary:
# Join Type	  Result
# inner	     common only
# left	     all left
# right	     all right
# outer	     everything
```

```python
# Merge on Different Column Names
pd.merge(df_students, df_marks, left_on="id", right_on="id")
```

```python
# Merge on Multiple Columns
import pandas as pd
df1= pd.read_json("student3.json")
df2=pd.read_json("marks.json")

pd.merge(df1, df2, on=["id","marks"])
```

```python
# Suffixes:If same column names exist

pd.merge(df1, df2, on="id", suffixes=("_left", "_right"))
```

```python
# join() Method:Alternative to merge.
# import pandas as pd
# df1= pd.read_json("student3.json")
# df2=pd.read_json("marks.json")

df1.join(df2.set_index("id"), on ="id" ,lsuffix="_left", rsuffix="_right")
# Join works on index.
```

```python
df = pd.merge(df_students, df_marks, on="id", how="left")

missing = df[df["marks"].isnull()]

print("All Data:\n", df)
print("Students without marks:\n", missing)
```
