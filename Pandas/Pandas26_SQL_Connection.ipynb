```python
# Use JOIN Queries with Pandas
query = """
SELECT TOP 20
    soh.SalesOrderID,
    soh.OrderDate,
    c.CustomerID
FROM Sales.SalesOrderHeader soh
JOIN Sales.Customer c
ON soh.CustomerID = c.CustomerID
"""

df = pd.read_sql(query, conn)

print(df.head())
```

```python
# SQL Aggregation + Pandas Visualization Prep
query = """
SELECT YEAR(OrderDate) AS OrderYear,
       COUNT(*) AS TotalOrders
FROM Sales.SalesOrderHeader
GROUP BY YEAR(OrderDate)
ORDER BY OrderYear
"""

df = pd.read_sql(query, conn)

print(df)
```

```python
# Chunking Large SQL Data
query = "SELECT * FROM Sales.SalesOrderDetail"

for chunk in pd.read_sql(query, conn, chunksize=1000):
    print(chunk.shape)
```

```python
# Data Cleaning After SQL Import
query = "SELECT TOP 50 FirstName, MiddleName FROM Person.Person"

df = pd.read_sql(query, conn)

df["MiddleName"] = df["MiddleName"].fillna("N/A")

print(df.head())
```

```python
# Creating Calculated Columns
query = """
SELECT TOP 20
SalesOrderID,
SubTotal,
TaxAmt
FROM Sales.SalesOrderHeader
"""

df = pd.read_sql(query, conn)

df["GrandTotal"] = df["SubTotal"] + df["TaxAmt"]

print(df.head())
```

```python
# Export SQL Data to Excel
df.to_excel("orders_report.xlsx", index=False)
```

```python
# Insert Filtered Data Back to SQL
high_orders = df[df["GrandTotal"] > 1000]

high_orders.to_sql(
    "HighValueOrders",
    engine,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT TOP 100 OrderDate, TotalDue
FROM Sales.SalesOrderHeader
"""

df = pd.read_sql(query, conn)

df["OrderDate"] = pd.to_datetime(df["OrderDate"])

report = df.groupby(df["OrderDate"].dt.year)["TotalDue"].sum()

print(report)
```
