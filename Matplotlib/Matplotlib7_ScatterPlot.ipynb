```python
# 2. What is a Scatter Plot?
# A scatter plot displays individual data points in a 2D plane:
# -X-axis → independent variable
# -Y-axis → dependent variable
# Each point represents a pair (x, y).
```

```python
# 3. Why Use Scatter Plots?
# Scatter plots are used to:
# ✔ Identify relationships
# ✔ Detect correlation
# ✔ Discover patterns
# ✔ Find outliers
# ✔ Visualize clustering
```

```python
# Clustering in Scatter Plot
# Clusters indicate:
# -grouping
# -segmentation
# -classification
```

```python
import numpy as np
import matplotlib.pyplot as plt
# -grouping
# Create 3 simple clusters of points
# Cluster 1: around (2, 8)
cluster1_x = [1.8, 2.1, 1.9, 2.2, 2.0, 1.7, 2.3, 1.9]
cluster1_y = [7.9, 8.2, 7.8, 8.1, 8.3, 7.7, 8.0, 8.2]

# Cluster 2: around (5, 5)
cluster2_x = [4.8, 5.1, 4.9, 5.2, 5.0, 4.7, 5.3, 5.1]
cluster2_y = [4.9, 5.2, 4.8, 5.1, 5.3, 4.7, 5.0, 4.8]

# Cluster 3: around (8, 2)
cluster3_x = [7.9, 8.2, 7.8, 8.1, 8.3, 7.7, 8.0, 8.2]
cluster3_y = [1.8, 2.1, 1.9, 2.2, 2.0, 1.7, 2.3, 1.9]

# Create scatter plot
plt.figure(figsize=(8, 6))

# Plot each cluster with different color
plt.scatter(cluster1_x, cluster1_y, color='red', s=80, alpha=0.7, label='Group A')
plt.scatter(cluster2_x, cluster2_y, color='blue', s=80, alpha=0.7, label='Group B')
plt.scatter(cluster3_x, cluster3_y, color='green', s=80, alpha=0.7, label='Group C')
# s = size of cluster balls (each point)
# Labels and title
plt.xlabel('X-axis (Feature 1)', fontsize=12)
plt.ylabel('Y-axis (Feature 2)', fontsize=12)
plt.title('Simple Scatter Plot Showing 3 Clusters', fontsize=14, fontweight='bold')

# Add legend
plt.legend()

# Add grid
plt.grid(True, alpha=0.3)

# Show plot
plt.show()
```

```python
#segmentation
import matplotlib.pyplot as plt

# Customer data: [Age, Spending Score]
# Group 1: Young, high spending
young_high = [[22, 85], [23, 88], [24, 82], [25, 90]]

# Group 2: Middle-aged, medium spending  
mid_medium = [[42, 55], [45, 52], [48, 58], [50, 53]]

# Group 3: Senior, low spending
senior_low = [[65, 25], [68, 22], [70, 28], [72, 20]]

# Plot the 3 segments
plt.figure(figsize=(8, 6))

# Extract x and y for each segment
young_x = [p[0] for p in young_high]
young_y = [p[1] for p in young_high]

# print(young_x) #[22, 23, 24, 25]
# print(young_y) #[85, 88, 82, 90]

mid_x = [p[0] for p in mid_medium]
mid_y = [p[1] for p in mid_medium]

# print(mid_x) #[42, 45, 48, 50]
# print(mid_y) #[55, 52, 58, 53]

senior_x = [p[0] for p in senior_low]
senior_y = [p[1] for p in senior_low]

# print(senior_x) #[65, 68, 70, 72]
# print(senior_y) #[25, 22, 28, 20]

# Create scatter plot with different colors
plt.scatter(young_x, young_y, color='red', s=100, label='Segment 1: Young High-Spenders')
plt.scatter(mid_x, mid_y, color='blue', s=100, label='Segment 2: Mid Medium-Spenders')
plt.scatter(senior_x, senior_y, color='green', s=100, label='Segment 3: Senior Low-Spenders')

plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Spending Score (0-100)', fontsize=12)
plt.title('Customer Segmentation: 3 Natural Groups', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Print segment summary
print("=== CUSTOMER SEGMENTS ===\n")
print(f"Segment 1 (Red):   Ages {young_x}, Spending {young_y}")
print(f"Segment 2 (Blue):  Ages {mid_x}, Spending {mid_y}")
print(f"Segment 3 (Green): Ages {senior_x}, Spending {senior_y}")
```

```python
#Classification
import matplotlib.pyplot as plt

# Three groups (each just 2 points)
group1_x = [1, 2]    # Group A
group1_y = [1, 2]

group2_x = [5, 6]    # Group B  
group2_y = [5, 6]

group3_x = [9, 10]   # Group C
group3_y = [9, 10]

# New point to classify
new_x = 3
new_y = 3

# Plot
plt.plot(group1_x, group1_y, 'ro', markersize=10, label='Group A')  # red circles:'ro' → red circles for Group A
plt.plot(group2_x, group2_y, 'bs', markersize=10, label='Group B')  # blue squares:'bs' → blue squares for Group B
plt.plot(group3_x, group3_y, 'g^', markersize=10, label='Group C')  # green triangles:'g^' → green triangles for Group C
plt.plot(new_x, new_y, 'k*', markersize=15, label='NEW → Classified as GROUP A')#'k*' → black star for the new point


plt.xlim(0, 11)
plt.ylim(0, 11)
plt.title('Simple Classification: New Point Joins Nearest Group')
plt.legend()
plt.grid(True)
plt.show()

print("NEW POINT (3,3) is CLOSEST to GROUP A → Classified as GROUP A")
```

```python
# Scatter Plot with Regression Line
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,6)
y = [2,4,5,4,5]

plt.scatter(x,y)

m, b= np.polyfit(x,y,1)
plt.plot(x, m*x + b, color='red')

plt.show()
```

```python
# Interpretation
# -Red line shows trend
# -Helps in prediction
# -Used in regression analysis
```
