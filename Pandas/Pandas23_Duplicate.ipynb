```python
# PART 1: WHAT ARE DUPLICATES?
# Why Duplicates Matter?
# Duplicates can seriously impact your analysis - they can double-count sales, overstate customer counts, and give wrong averages.
```

```python
import pandas as pd
import numpy as np

print("="*70)
print("WHY DUPLICATES ARE A PROBLEM")
print("="*70)

# Create data with duplicates
sales_data = pd.DataFrame({
    'Order_ID': [1001, 1002, 1003, 1001, 1004, 1002],
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Monitor', 'Mouse'],
    'Quantity': [2, 5, 3, 2, 1, 5],
    'Price': [1000, 25, 50, 1000, 300, 25]
})

print("Sales Data WITH Duplicates:")
print(sales_data)
```

```python
print("\n PROBLEM 1: Counting Orders (Wrong!)")
print(f"Order count with duplicates: {len(sales_data)}")
print(f"Unique order count: {sales_data['Order_ID'].nunique()}")
```

```python
print("\n PROBLEM 2: Total Quantity (Wrong!)")
print(f"Total quantity with duplicates: {sales_data['Quantity'].sum()}")
correct_qty = sales_data.drop_duplicates(subset=['Order_ID'])['Quantity'].sum()
print(f"Correct total quantity: {correct_qty}")
```

```python
print("\n PROBLEM 3: Average Price (Wrong!)")
print(f"Average price with duplicates: ${sales_data['Price'].mean():.2f}")
correct_avg = sales_data.drop_duplicates(subset=['Order_ID'])['Price'].mean()
print(f"Correct average price: ${correct_avg:.2f}")

print("\n SOLUTION: Remove duplicates for accurate analysis!")
```

```python
# PART 2: IDENTIFYING DUPLICATES
```

```python
# 2.1 Using duplicated() Method
print("="*70)
print("IDENTIFYING DUPLICATES")
print("="*70)

# Create dataset with various duplicates
df = pd.DataFrame({
    'ID': [1, 2, 3, 1, 4, 2, 5, 3],
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Bob', 'Emma', 'Charlie'],
    'Score': [85, 92, 78, 85, 88, 92, 95, 78]
})

print("Original DataFrame:")
print(df)
```

```python
# 1. Find duplicate rows (all columns must match)
print("\n Find duplicate rows (exact duplicates):")
duplicates = df.duplicated()
print(duplicates)

print("\n Duplicate rows (marked True):")
print(df[duplicates])
```

```python
# 2. Show all duplicates including first occurrence
print("\n Show all duplicates including first (keep=False):")
all_duplicates = df.duplicated(keep=False)
print(df[all_duplicates].sort_values('ID'))
```

```python
# 3. Count duplicates
print("\n Duplicate Statistics:")
print(f"Total rows: {len(df)}")
print(f"Duplicate rows: {duplicates.sum()}")
print(f"Unique rows: {len(df) - duplicates.sum()}")
```

```python
# 4. Check which rows are duplicated
print("\n Which specific rows are duplicates?")
df['Is_Duplicate'] = df.duplicated(keep='first')
print(df)
```

```python
# 4. Check which rows are duplicated
print("\n Which specific rows are duplicates?")
df['Is_Duplicate'] = df.duplicated(keep='last')
print(df)
```

```python
# 3.1 Basic Duplicate Removal
```

```python
print("="*70)
print("REMOVING DUPLICATES")
print("="*70)

# Create data with duplicates
df_dupes = pd.DataFrame({
    'Product': ['Laptop', 'Mouse','Printer', 'Keyboard', 'Laptop', 'Mouse', 'Monitor', 'Keyboard','Printer','VGA'],
    'Price': [100, 25, 50, 1000, 125, 300, 150, 100, 50,500],
    'Stock': [11, 150, 75, 15, 160, 8, 55, 20, 75, 10]
})

print("Data with Duplicates:")
print(df_dupes)
```

```python
# 1. Remove all duplicate rows
print("Remove all duplicate rows (keep first):")
df_no_dupes = df_dupes.drop_duplicates()
print(df_no_dupes)
print(f"Rows removed: {len(df_dupes) - len(df_no_dupes)}")
```

```python
# 2. Remove duplicates and modify original
print("Remove duplicates in-place:")
df_copy = df_dupes.copy()
df_copy.drop_duplicates(inplace=True)
print(df_copy)
```

```python
# 3. See what was removed
print("Identify which rows will be removed:")
df_dupes['To_Remove'] = df_dupes.duplicated(keep='first')
print(df_dupes)
```

```python
print("="*70)
print("REMOVING DUPLICATES BASED ON SPECIFIC COLUMNS")
print("="*70)

# Create dataset
orders = pd.DataFrame({
    'Order_ID': [1001, 1002, 1003, 1001, 1004, 1002, 1005],
    'Customer': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Bob', 'Emma'],
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Monitor', 'Mouse', 'Tablet'],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-01', 
             '2024-01-04', '2024-01-02', '2024-01-05']
})

print("Orders Data:")
print(orders)
```

```python
# 1. Remove duplicates based on Order_ID only
print("Keep only unique Order_IDs:")
unique_orders = orders.drop_duplicates(subset=['Order_ID'])
print(unique_orders)
print(f"Original: {len(orders)}, After: {len(unique_orders)}")
```

```python
# 2. Remove duplicates based on Customer and Product
print("Each customer can order same product only once:")
unique_customer_product = orders.drop_duplicates(subset=['Customer', 'Product'])
print(unique_customer_product)
```

```python
# 3. Remove duplicates based on multiple columns
print("Unique combinations of Customer, Product, Date:")
unique_combinations = orders.drop_duplicates(subset=['Customer', 'Product', 'Date'])
print(unique_combinations)
```

```python
# PART 4: KEEPING SPECIFIC OCCURRENCES
```

```python
print("="*70)
print("CONTROLLING WHICH DUPLICATE TO KEEP")
print("="*70)

# Create dataset
employee_data = pd.DataFrame({
    'Employee_ID': [101, 102, 101, 103, 102, 104, 101],
    'Name': ['John', 'Jane', 'John', 'Bob', 'Jane', 'Alice', 'John'],
    'Salary': [50000, 60000, 55000, 45000, 62000, 70000, 58000],
    'Date': ['2023-01-01', '2023-01-01', '2024-01-01', '2023-01-01', 
             '2024-01-01', '2023-01-01', '2025-01-01']
})

print("Employee Records (Multiple entries per employee):")
print(employee_data)
```

```python
# 1. Keep first occurrence (default)
print("Keep FIRST occurrence:")
keep_first = employee_data.drop_duplicates(subset=['Employee_ID'], keep='first')
print(keep_first)
```

```python
# 2. Keep last occurrence
print("Keep LAST occurrence:")
keep_last = employee_data.drop_duplicates(subset=['Employee_ID'], keep='last')
print(keep_last)
```

```python
# 3. Remove all duplicates (keep none)
print("Remove ALL duplicates (keep=False):")
remove_all = employee_data.drop_duplicates(subset=['Employee_ID'], keep=False)
print(remove_all)
```

```python
# 4. Practical example: Keep latest record based on date
print("Practical: Keep latest record for each employee")

# Sort by date first (newest first)
employee_sorted = employee_data.sort_values('Date', ascending=False)
latest_records = employee_sorted.drop_duplicates(subset=['Employee_ID'], keep='first')
print(latest_records)
```

```python
# 5.1 Finding Duplicate Patterns
print("="*70)
print("ANALYZING DUPLICATE PATTERNS")
print("="*70)

# Create rich dataset for analysis
sales = pd.DataFrame({
    'Store': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Laptop', 'Keyboard', 'Mouse'],
    'Sales': [1000, 50, 1000, 75, 50, 1200, 75, 55],
    'Month': ['Jan', 'Jan', 'Feb', 'Jan', 'Feb', 'Mar', 'Feb', 'Mar']
})

print("Sales Data:")
print(sales)
```

```python
# 1. Find which columns have duplicates
print("Duplicate analysis by column:")
for col in sales.columns:
    duplicates = sales.duplicated(subset=[col]).sum()
    print(f"  {col}: {duplicates} duplicate values")
```

```python
# 5.1 Finding Duplicate Patterns
print("="*70)
print("ANALYZING DUPLICATE PATTERNS")
print("="*70)

# Create rich dataset for analysis
sales = pd.DataFrame({
    'Store': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B','D'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Laptop', 'Keyboard', 'Mouse','Joystick'],
    'Sales': [1000, 50, 1000, 75, 50, 1200, 75, 55, 85],
    'Month': ['Jan', 'Jan', 'Feb', 'Jan', 'Feb', 'Mar', 'Feb', 'Mar', 'June']
})

print("Sales Data:")
print(sales)
```

```python
# 2. Find duplicate rows and see the pattern
print("Duplicate rows analysis:")
remove_all = sales.drop_duplicates(subset=['Store'], keep=False)
print(remove_all)

sales['Is_Duplicate'] = sales.duplicated(subset=['Store'],keep=False)
print(sales['Is_Duplicate'])
duplicate_groups = sales[sales['Is_Duplicate']].sort_values(['Store', 'Product'])
print(duplicate_groups)
```

```python
# 3. Count how many times each combination appears
print("Count frequency of each row:")
frequency = sales.groupby(list(sales.columns[:-1])).size().reset_index(name='Count')
print(frequency)
```

```python
# 4. Find perfect duplicates (all columns match)
print("Perfect duplicates (all columns match):")
perfect_dupes = sales[sales.duplicated(keep=False)]
print(perfect_dupes.sort_values(['Store', 'Product']))
```

```python
# 5.2 Visualizing Duplicate Statistics
print("="*70)
print("DUPLICATE STATISTICS")
print("="*70)
# seed(): it will allow same repeated value
# Create larger dataset
np.random.seed(42)
large_data = pd.DataFrame({
    'ID': np.random.choice([1, 2, 3, 4, 5], 100),
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value': np.random.randint(1, 100, 100)
})

print("Dataset Statistics:")
print(f"Total rows: {len(large_data)}")

print(f"Unique rows: {len(large_data.drop_duplicates())}")
print(f"Duplicate rows: {len(large_data) - len(large_data.drop_duplicates())}")
print(f"Duplicate percentage: {(len(large_data) - len(large_data.drop_duplicates())) / len(large_data) * 100:.1f}%")
```

```python
# Duplicate breakdown by ID
print("Duplicate count by ID:")
id_counts = large_data['ID'].value_counts()
print(id_counts)
print(f"\nIDs with duplicates: {(id_counts > 1).sum()}")
print(f"IDs without duplicates: {(id_counts == 1).sum()}")
```

```python
# Find most common duplicate patterns
print("Most common duplicate patterns:")
pattern_counts = large_data.groupby(['ID', 'Category']).size().reset_index(name='Count')
print(pattern_counts[pattern_counts['Count'] > 1].sort_values('Count', ascending=False))
```
