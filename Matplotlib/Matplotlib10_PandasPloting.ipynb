```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Day": [1,2,3,4,5],
    "Sales": [100,120,130,150,170]
}

df = pd.DataFrame(data)
print(df)
df.set_index("Day", inplace=True)
df.plot(title="Daily Sales")

# ax = df.plot(title="Daily Sales")

# ax.legend(['Analysis'])
# plt.show()
```

```python
# Multiple Columns Plot
data = {
    "Day": [1,2,3,4,5],
    "Product_A": [100,120,130,150,170],
    "Product_B": [90,110,125,140,160]
}

df = pd.DataFrame(data).set_index("Day")

df.plot(title="Sales Comparison")
```

```python
# | Plot      | Code                                    |
# | --------- | --------------------------------------- |
# | Line      | `df.plot()`                             |
# | Bar       | `df.plot(kind='bar')`                   |
# | Histogram | `df.plot(kind='hist')`                  |
# | Scatter   | `df.plot(kind='scatter', x=..., y=...)` |
```

```python
# Case Study 1: Student Performance Analysis
# Dataset
data = {
    "Student": ["A","B","C","D","E"],
    "Marks": [75,85,60,90,70]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Bar Plot
df.plot(kind='bar', x='Student', y='Marks',
        title="Student Performance",
        color='green')
```

```python
# Case Study 1: Student Performance Analysis
# Dataset
data = {
    "RollNo": [101,102,103,104,105],
    "Marks": [75,85,60,90,70]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Bar Plot
df.plot(kind='bar', x='RollNo', y='Marks',
        title="Student Performance",
        color='green')

# plt.xlabel("RollNo", fontsize=12)
# plt.ylabel("Marks", fontsize=12)
```

```python
# Interpretation
# -Student D highest
# -Student C lowest
# -Easy categorical comparison
```

```python
# Case Study 2: Time Series (Stock Trend)
# Dataset
data = {
    "Date": ["2024-01","2024-02","2024-03","2024-04"],
    "Price": [100,120,110,130]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Plot
df.set_index("Date", inplace=True)
df.plot(kind='line',
        title="Stock Price Trend",
        color='red')
```

```python
# Interpretation
# -Upward trend overall
# -Dip in March
# -Useful for financial analysis
```

```python
# Case Study 3: Distribution Analysis
import numpy as np

data = np.random.normal(70,10,100)

df = pd.DataFrame({"Marks": data})

print(df)

df.plot(kind='hist', bins=20, edgecolor="black",title="Marks Distribution")
```

```python
# Interpretation
# -Normal-like distribution
# -Most students near average
```

```python
# Case Study 4: Relationship Analysis
data = {
    "Hours": [1,2,3,4,5],
    "Marks": [40,50,60,70,80]
}

df = pd.DataFrame(data)

print(df)

df.plot(kind='scatter', x='Hours', y='Marks',
        title="Study vs Marks")
```

```python
# Interpretation
# -Positive correlation
# -Useful for ML feature analysis
```

```python
# Customizing Pandas Plots
df.plot(
    kind='line',
    figsize=(8,5),
    color=['blue','red'],
    marker='o',
    title="Styled Plot"
)
```

```python
# Accessing Matplotlib from Pandas
import matplotlib.pyplot as plt

ax = df.plot()
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")

plt.show()
```

```python
# Subplots in Pandas
df.plot(subplots=True, layout=(2,1), figsize=(8,6))
```

```python
# Real Dataset Example
# Simulated Sales Dataset
import pandas as pd

data = {
    "Month": ["Jan","Feb","Mar","Apr","May"],
    "Revenue": [200,220,250,270,300],
    "Profit": [50,60,70,65,80]
}

df = pd.DataFrame(data).set_index("Month")
print(df)

df.plot(title="Business Performance")
```
