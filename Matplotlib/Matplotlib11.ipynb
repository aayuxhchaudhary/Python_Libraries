```python
# Grouped Bar Plot
# When DataFrame contains multiple numerical columns:
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Branch": ["CSE","IT","ECE"],
    "Boys": [80,50,45],
    "Girls": [40,40,35]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Plot
df.plot(
    kind='bar',
    x='Branch',
    figsize=(8,5)
)

plt.title("Gender Distribution")
plt.ylabel("Students")

plt.show()
```

```python
# What Happens Internally?
# Pandas automatically:
# -creates grouped bars
# -assigns colors
# -generates legend
```

```python
# Interpretation
# Students must answer:
# -Which branch has highest girls count?
# -Which branch has highest total students?
# -Compare ratios
```

```python
df.plot(
    kind='bar',
    x='Branch',
    stacked=True,
    figsize=(8,5)
)

plt.title("Stacked Gender Distribution")

plt.show()
```

```python
# Interpretation
# -Total height = total students
# -Colors = composition
```

```python
# Horizontal DataFrame Bar Plot
df.plot(
    kind='barh',
    x='Branch',
    figsize=(8,5)
)
```

```python
# Real Case Study: Sales Analytics
data = {
    "Month": ["Jan","Feb","Mar","Apr"],
    "Laptop": [50,60,70,80],
    "Mobile": [80,90,85,95]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.plot(
    kind='bar',
    x='Month',
    figsize=(8,5)
)

plt.title("Monthly Product Sales")
plt.ylabel("Units Sold")

plt.show()
```

```python
data = {
    "Model": ["SVM","RF","KNN","NN"],
    "Accuracy": [85,90,80,92]
}

df = pd.DataFrame(data)
print(df)

df.plot(
    kind='bar',
    x='Model',
    y='Accuracy',
    legend=False,
    color='orange'
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")

plt.show()
```

```python
df.plot(
    kind='bar',
    color=['blue','green']
)
```

```python
# Sorting Data Before Plotting
df.sort_values(by='Accuracy').plot(
    kind='bar',
    x='Model',
    y='Accuracy'
)
```
