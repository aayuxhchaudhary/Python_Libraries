```python
import pandas as pd
import numpy as np

# Create a sample dataset of students
student_data = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice Johnson', 'Bob Smith', 'Carol Davis', 'David Wilson', 'Emma Brown'],
    'Age': [20, 21, 19, 22, 20],
    'Grade': ['A', 'B', 'A+', 'C', 'B+'],
    'Marks': [85.5, 72.3, 92.0, 65.5, 78.8],
    'Attendance': [95, 88, 98, 75, 90]
}

df = pd.DataFrame(student_data)
print("Original DataFrame:")
print(df)
```

```python
# 1. Basic Excel Export

# Simple export to Excel
df.to_excel('students.xlsx', index=False, sheet_name='Student Data')
print("Created: students.xlsx with sheet 'Student Data'")
```

```python
# 2. Multiple Sheets in One Excel File

# Create multiple dataframes
grades_summary = df.groupby('Grade').agg({
    'Marks': ['mean', 'min', 'max', 'count']
}).round(2)

attendance_summary = pd.DataFrame({
    'Attendance_Range': ['<80%', '80-90%', '>90%'],
    'Student_Count': [1, 1, 3],
    'Average_Marks': [65.5, 78.8, 85.27]
})

# Export to multiple sheets
with pd.ExcelWriter('complete_report.xlsx', engine='openpyxl') as writer:
    # Sheet 1: Raw data
    df.to_excel(writer, sheet_name='Raw_Data', index=False)
    
    # Sheet 2: Grades summary
    grades_summary.to_excel(writer, sheet_name='Grade_Analysis')
    
    # Sheet 3: Attendance analysis
    attendance_summary.to_excel(writer, sheet_name='Attendance_Analysis', index=False)
    
    # Sheet 4: Top performers
    top_students = df.nlargest(3, 'Marks')[['Name', 'Marks', 'Grade']]
    top_students.to_excel(writer, sheet_name='Top_Performers', index=False)

print("Created: complete_report.xlsx with 4 sheets")
```

```python
#3. Starting at Specific Position

# Export starting from specific row and column
df.to_excel('positioned_data.xlsx', 
            startrow=2,    # Start at row 3 (0-indexed)
            startcol=1,    # Start at column B
            index=False)
print("Created: positioned_data.xlsx (data starts at B3)")
```

```python
#4. Handling Multi-Index DataFrames

# Create multi-index DataFrame
multi_index_df = df.set_index(['Grade', 'Age']).sort_index()
print("Multi-Index DataFrame:")
print(multi_index_df.head())
print("\n")

# Export with merged cells
multi_index_df.to_excel('multi_index.xlsx', merge_cells=True)
print("Created: multi_index.xlsx with merged cells for index")
```

```python
#5. Adding Charts to Excel
import xlsxwriter

# Create Excel with chart
with pd.ExcelWriter('chart_report.xlsx', engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='Data', index=False)
    
    # Get workbook and worksheet
    workbook = writer.book
    worksheet = writer.sheets['Data']
    
    # Create a chart
    chart = workbook.add_chart({'type': 'column'})
    
    # Configure the chart
    chart.add_series({
        'name': 'Student Marks',
        'categories': ['Data', 1, 1, 5, 1],  # Names in column B
        'values': ['Data', 1, 5, 5, 5],      # Marks in column F
    })
    
    # Insert chart
    worksheet.insert_chart('H2', chart)

print("Created: chart_report.xlsx with embedded chart")
```
