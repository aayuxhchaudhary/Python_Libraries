```python
# PART 1: INTRODUCTION TO ELEMENT-WISE OPERATIONS
# Why Do We Need map(), apply(), and applymap()?
# Sometimes you need to apply custom logic to every element in your data. These methods allow you to do that efficiently.
```

```python
import pandas as pd
import numpy as np

print("="*70)
print("WHY WE NEED map(), apply(), and applymap()")
print("="*70)

# Create sample data
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Math_Score': [85, 92, 78, 88, 95],
    'English_Score': [90, 88, 85, 92, 87],
    'Attendance': [95, 88, 92, 85, 98]
})

print("Original DataFrame:")
print(df)

print("\n PROBLEM: What if we want to:")
print("1. Convert all scores to letter grades (A, B, C)?")
print("2. Categorize attendance as 'Good', 'Average', 'Poor'?")
print("3. Clean text data in names?")
print("4. Apply complex calculations to each row?")

print("\n SOLUTION: map(), apply(), and applymap()!")
print("• map() - Transform values in a Series using a dictionary or function")
print("• apply() - Apply function to Series or DataFrame (row/column-wise)")
print("• applymap() - Apply function element-wise to entire DataFrame")
```

```python
# PART 2: map() - Transform Series Values
# What is map()?
# map() is used to substitute each value in a Series with another value. 
# It's perfect for categorical mapping and simple transformations.
```

```python
print("="*70)
print("UNDERSTANDING map() - Series Only")
print("="*70)

# Create sample Series
scores = pd.Series([85, 92, 78, 88, 95])
print("Original Scores Series:")
print(scores)
```

```python
# Example 1: Using a dictionary for mapping
print("\n Example 1: Converting numbers to letter grades using dictionary")
grade_map = {
    85: 'B+',
    92: 'A-',
    78: 'C+',
    88: 'B+',
    95: 'A'
}
# print(grade_map[85])
mapped_grades = scores.map(grade_map)
print("Mapped Grades:")
print(mapped_grades)
```

```python
# Example 2: Using a function for mapping
print("\n Example 2: Using a function to classify scores")
def score_category(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 80:
        return 'Good'
    elif score >= 70:
        return 'Average'
    else:
        return 'Needs Improvement'

category_map = scores.map(score_category)
print("Score Categories:")
print(category_map)
```

```python
# Example 3: Practical - Grade students based on marks
print("\n Example 3: Real-world - Student Grading")
students = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Percentage': [85, 92, 78, 88, 95]
})

# Create grade mapping dictionary
grade_dict = {
    85: 'B+', 92: 'A-', 78: 'C+', 88: 'B+', 95: 'A'
}
students['Grade'] = students['Percentage'].map(grade_dict)
print(students)
```

```python
# Example 4: Handling missing mappings with na_action
print("\n Example 4: Handling values not in mapping dictionary")
scores_with_new = pd.Series([85, 92, 100, 78, 88, 95])
print("New scores (including 100 not in dictionary):")
print(scores_with_new)
```

```python
mapped_with_default = scores_with_new.map(grade_map)
print("\nMapped (100 becomes NaN):")
print(mapped_with_default)
```

```python
# Fill missing values
mapped_filled = scores_with_new.map(grade_map).fillna('A+')
print("\nMapped with fillna (100 becomes A+):")
print(mapped_filled)
```

```python
# Example 5: Using map() for data cleaning
print("\n Example 5: Data Cleaning with map()")
messy_names = pd.Series(['alice', 'BOB', 'Charlie', 'david', 'EMMA'])
print("Messy names:")
print(messy_names)

cleaned_names = messy_names.map(lambda x: x.capitalize())
print("\nCleaned names (capitalized):")
print(cleaned_names)
```

```python
# Example 6: Boolean mapping
print("\n Example 6: Creating boolean flags with map()")
scores_series = pd.Series([85, 92, 78, 88, 95])
is_passing = scores_series.map(lambda x: x >= 70)
print("Is passing (>=70)?")
print(is_passing)
```

```python
# PART 3: apply() - Most Powerful & Flexible
# What is apply()?
# apply() is the most versatile method. It can:
# Apply functions to Series (element-wise)
# Apply functions to DataFrame (row-wise or column-wise)
```

```python
print("="*70)
print("UNDERSTANDING apply() - Series & DataFrame")
print("="*70)

# Create sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Math': [85, 92, 78, 88, 95],
    'Science': [88, 85, 90, 87, 92],
    'English': [90, 88, 85, 92, 87]
})

print("Original DataFrame:")
print(df)
```

```python
# SECTION 1: apply() on Series (Column-wise)
print("\n" + "="*50)
print("apply() ON SERIES (Single Column)")
print("="*50)

# Example 1: Apply function to a single column
print("\n Example 1: Calculate grade category for Math scores")
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

df['Math_Grade'] = df['Math'].apply(get_grade)
print(df[['Name', 'Math', 'Math_Grade']])
```

```python
# Example 2: Using lambda with apply()
print("\n Example 2: Using lambda to add 5 bonus points")
df['Math_Bonus'] = df['Math'].apply(lambda x: x + 5)
print(df[['Name', 'Math', 'Math_Bonus']])
```

```python
# Example 3: Multiple operations in one apply()
print("\n Example 3: Complex transformation")
def analyze_score(score):
    return {
        'original': score,
        'bonus_5': score + 5,
        'grade': get_grade(score),
        'is_passing': score >= 70
    }

analysis = df['Math'].apply(analyze_score)
print("Analysis for first student:")
print(analysis.iloc[0])
```

```python
# SECTION 2: apply() on DataFrame (Row-wise)
print("\n" + "="*50)
print(" apply() ON DATAFRAME (Row-wise)")
print("="*50)

# Example 4: Calculate average across subjects (row-wise)
print("\n Example 4: Calculate average score for each student")
df['Average'] = df.apply(lambda row: (row['Math'] + row['Science'] + row['English']) / 3, axis=1)
print(df[['Name', 'Math', 'Science', 'English', 'Average']])
```

```python
# Example 5: Determine overall performance category
print("\n Example 5: Overall performance category")
def overall_grade(row):
    avg = (row['Math'] + row['Science'] + row['English']) / 3
    if avg >= 90:
        return 'Excellent'
    elif avg >= 80:
        return 'Good'
    elif avg >= 70:
        return 'Average'
    else:
        return 'Poor'

df['Performance'] = df.apply(overall_grade, axis=1)
print(df[['Name', 'Average', 'Performance']])
```

```python
# Example 6: Find best subject for each student
print("\n Example 6: Find best subject for each student")
def best_subject(row):
    scores = {'Math': row['Math'], 'Science': row['Science'], 'English': row['English']}
    return max(scores, key=scores.get)

df['Best_Subject'] = df.apply(best_subject, axis=1)
print(df[['Name', 'Math', 'Science', 'English', 'Best_Subject']])
```

```python
# SECTION 3: apply() on DataFrame (Column-wise)
print("\n" + "="*50)
print(" apply() ON DATAFRAME (Column-wise)")
print("="*50)

# Example 7: Apply to multiple columns at once
print("\n Example 7: Calculate statistics for all numeric columns")
stats = df[['Math', 'Science', 'English']].apply(lambda col: col.max() - col.min())
print("Range (Max - Min) for each subject:")
print(stats)
```

```python
# Example 8: Custom column transformation
print("\n Example 8: Normalize scores to 0-100 scale")
def normalize(column):
    return (column - column.min()) / (column.max() - column.min()) * 100

normalized_scores = df[['Math', 'Science', 'English']].apply(normalize)
print("Normalized Scores (0-100 scale):")
print(normalized_scores.round(2))
```

```python
# STEP 1: Create Average column (IMPORTANT)
# df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# SECTION 4: Advanced apply() Examples
print("\n" + "="*50)
print(" ADVANCED apply() EXAMPLES")
print("="*50)

print("\nExample 9: Flag students who need improvement")

# STEP 2: Define function
def needs_improvement(row):
    if row['Math'] < 70 or row['Science'] < 70 or row['English'] < 70:
        return 'Needs Improvement'
    elif row['Average'] < 90:
        return 'Below Average'
    else:
        return 'On Track'

# STEP 3: Apply function
df['Status'] = df.apply(needs_improvement, axis=1)

# STEP 4: Final Output
print(df[['Name', 'Average', 'Status']])
```

```python
# Example 10: Create detailed report
print("\n Example 10: Generate student report")
def generate_report(row):
    report = f"""
    Student: {row['Name']}
    Math: {row['Math']} ({get_grade(row['Math'])})
    Science: {row['Science']} ({get_grade(row['Science'])})
    English: {row['English']} ({get_grade(row['English'])})
    Average: {row['Average']:.1f}
    Performance: {row['Performance']}
    Best Subject: {row['Best_Subject']}
    """
    return report

df['Report'] = df.apply(generate_report, axis=1)
print("Sample Report for Alice:")
print(df[df['Math'] == 85 ]['Report'].iloc[1])
```

```python
# PART 4: applymap() - Element-wise DataFrame Operations
# What is applymap()?
# applymap() applies a function to every single element in the entire DataFrame.
```

```python
print("="*70)
print("UNDERSTANDING applymap() - Element-wise DataFrame")
print("="*70)

# Create sample DataFrame
df_numbers = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})

print(" Original DataFrame:")
print(df_numbers)
```

```python
# Example 1: Basic applymap - square every number
print("\n Example 1: Square every number in DataFrame")
squared = df_numbers.applymap(lambda x: x ** 2)
print(squared)
```

```python
# Example 2: Format numbers as currency
print("\n Example 2: Format every number as currency")
formatted = df_numbers.applymap(lambda x: f"${x:,.2f}")
print(formatted)
```

```python
# Example 3: Classify each number
print("\n Example 3: Classify each number (Small/Medium/Large)")
def classify_number(x):
    if x < 20:
        return 'Small'
    elif x < 100:
        return 'Medium'
    else:
        return 'Large'

classified = df_numbers.applymap(classify_number)
print(classified)
```

```python
# Example 4: Data cleaning with applymap
print("\n Example 4: Clean messy text data")
messy_text = pd.DataFrame({
    'Name': ['  Alice  ', 'BOB', 'Charlie', '  david  ', 'EMMA  '],
    'City': ['  NY  ', 'LA', 'CHICAGO', '  houston  ', 'MIAMI  ']
})
print("Messy DataFrame:")
print(messy_text)

cleaned_text = messy_text.applymap(lambda x: x.strip().capitalize())
print("\nCleaned DataFrame:")
print(cleaned_text)
```

```python
# Example 5: Applymap with numeric transformation
print("\n Example 5: Apply multiple transformations")
def transform(x):
    if isinstance(x,(int)):
        return x * 2  # Double numbers
    else:
        return str(x).upper()  # Uppercase text

df_mixed = pd.DataFrame({
    'Numbers': [1, 2, 3, 4, 5],
    'Text': ['a', 'b', 'c', 'd', 'e'],
    'Mixed': [10, 'x', 20, 'y', 30]
})

transformed = df_mixed.applymap(transform)
print("Original:")
print(df_mixed)
print("\nTransformed:")
print(transformed)
```

```python
# Example 5: Applymap with numeric transformation
print("\n Example 5: Apply multiple transformations")
def transform(x):
    if isinstance(x,(int,float)):
        return x * 2  # Double numbers
    else:
        return str(x).upper()  # Uppercase text

df_mixed = pd.DataFrame({
    'Numbers': [1, 2, 3, 4, 5],
    'Text': ['a', 'b', 'c', 'd', 'e'],
    'Mixed': [10.12, 'x', 20.15, 'y', 30]
})

transformed = df_mixed.applymap(transform)
print("Original:")
print(df_mixed)
print("\nTransformed:")
print(transformed)
```

```python
# Example 6: Highlight specific values
print("\n Example 6: Mark values above threshold")
def highlight_high_values(x):
    if isinstance(x, (int, float)) and x > 30:
        return f"★ {x} ★"
    return x

highlighted = df_numbers.applymap(highlight_high_values)
print(highlighted)
```

```python
# Example 7: Round all numbers
print("\n Example 7: Round all numbers to 2 decimal places")
df_decimals = pd.DataFrame({
    'Price': [19.991, 29.949, 39.995, 49.9980],
    'Tax': [1.9920, 2.9956, 3.9299, 4.9999]
})
rounded = df_decimals.applymap(lambda x: round(x, 2))
print("Original:")
print(df_decimals)
print("\nRounded:")
print(rounded)
```

```python
# PART 5: COMPARISON OF ALL THREE METHODS
print("="*70)
print("COMPARISON: map() vs apply() vs applymap()")
print("="*70)

# Create comprehensive comparison
comparison_df = pd.DataFrame({
    'Feature': ['map()', 'apply()', 'applymap()'],
    'Works On': ['Series only', 'Series & DataFrame', 'DataFrame only'],
    'Return Type': ['Series', 'Series or DataFrame', 'DataFrame'],
    'Use Case': ['Value mapping', 'Complex functions', 'Element-wise ops'],
    'Speed': ['Fast', 'Medium', 'Slow'],
    'Can Use dict': ['Yes', 'No', 'No'],
    'Axis Parameter': ['No', 'Yes (0/1)', 'No']
})

print(comparison_df.to_string(index=False))

print("\n" + "="*70)
print(" WHEN TO USE WHICH?")
print("="*70)

print("""
 USE map() WHEN:
   • You have a dictionary for value mapping
   • You're working with a single column (Series)
   • You need simple value replacement
   • Example: Replace 'M' with 'Male', 'F' with 'Female'

 USE apply() WHEN:
   • You need row-wise or column-wise operations
   • You have complex functions that return multiple values
   • You need to access multiple columns at once
   • Example: Calculate average of multiple columns for each row

 USE applymap() WHEN:
   • You need to apply the same function to every element
   • You're formatting or cleaning entire DataFrame
   • You need element-wise type conversion
   • Example: Format all numbers as currency, clean all text
""")
```

```python
# Live demonstration of differences
print("\n" + "="*70)
print("LIVE DEMONSTRATION: Same task with different methods")
print("="*70)

demo_df = pd.DataFrame({
    'Score': [85, 92, 78, 88, 95],
    'Bonus': [5, 3, 4, 2, 5]
})

print("Original DataFrame:")
print(demo_df)
```

```python
# Using map() - can only work on one column
print("\n 1️. Using map() on 'Score' column:")
demo_df['Score_Map'] = demo_df['Score'].map(lambda x: x + 5)
print(demo_df[['Score', 'Score_Map']])
```

```python
# Using apply() - can work on rows
print("\n 2. Using apply() row-wise to sum both columns:")
demo_df['Total_Apply'] = demo_df.apply(lambda row: row['Score'] + row['Bonus'], axis=1)
print(demo_df[['Score', 'Bonus', 'Total_Apply']])
```

```python
# Using applymap() - applies to every element
print("\n 3. Using applymap() to format every number:")
formatted_df = demo_df.applymap(lambda x: f"{x:.0f} pts")
print(formatted_df)
```

```python
 # PART 6: PERFORMANCE CONSIDERATIONS
print("="*70)
print("PERFORMANCE COMPARISON")
print("="*70)

# Create large dataset for performance test
import time

large_df = pd.DataFrame({
    'A': np.random.randint(1, 100, 10000),
    'B': np.random.randint(1, 100, 10000),
    'C': np.random.randint(1, 100, 10000)
})

print("Testing with 10,000 rows...")
```

```python
# Test 1: Vectorized operation (fastest)
start = time.time()
result_vectorized = large_df['A'] * 2
vectorized_time = time.time() - start
print(vectorized_time)
```

```python
# Test 2: apply() with lambda
start = time.time()
result_apply = large_df['A'].apply(lambda x: x * 2)
apply_time = time.time() - start
print(apply_time)
```

```python
# Test 3: map() with lambda
start = time.time()
result_map = large_df['A'].map(lambda x: x * 2)
map_time = time.time() - start

print(f"\n Performance Results:")
print(f"Vectorized (fastest): {vectorized_time:.4f} seconds")
print(f"apply(): {apply_time:.4f} seconds")
print(f"map(): {map_time:.4f} seconds")

print("\n PERFORMANCE TIPS:")
print("1. Use vectorized operations when possible (fastest)")
print("2. Use map() for dictionary lookups (fast)")
print("3. Use apply() for complex row/column operations")
print("4. Use applymap() for element-wise formatting (slowest but flexible)")
print("5. Avoid apply() when vectorized operations exist")
```
