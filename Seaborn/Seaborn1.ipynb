```python
# Seaborn with Pandas DataFrame
import pandas as pd

data = {
    "Month": ["Jan","Feb","Mar","Apr","May"],
    "Sales": [200,220,250,270,300]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Plot
sns.lineplot(
    data=df,#data frame mae kra data upper phir abb vapis dataframeko 
    x="Month",
    y="Sales"
)

plt.title("Monthly Sales Trend")

plt.show()
```

```python
# Load Dataset
tips = sns.load_dataset("tips")

print(tips.head())
```

```python
# Understanding the Tips Dataset
# Columns include:
# | Column     | Meaning           |
# | ---------- | ----------------- |
# | total_bill | restaurant bill   |
# | tip        | tip amount        |
# | sex        | gender            |
# | smoker     | smoker/non-smoker |
# | day        | weekday           |
# | time       | lunch/dinner      |
```

```python
# First Real Dataset Plot
tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.title("Bill vs Tip")

plt.show()
```

```python
# 20. Interpretation
# Students should analyze:
# -larger bills often lead to larger tips
# -positive relationship
# -some variability exists
```

```python
# Adding Categories using Hue
# -One powerful Seaborn feature.
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"#issa x walo or y walo ko alg alg color mila ga jase upper sabko blue mila par yha nhi mila ga
)

plt.show()
```

```python
# 22. What is Hue?
# hue adds color encoding for categories.
# Example:
# -male vs female
# -smoker vs non-smoker
```
