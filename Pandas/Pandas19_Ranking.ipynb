```python
import pandas as pd
import numpy as np

# Create sample dataset
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace'],
    'Math_Score': [85, 92, 78, 92, 88, 85, 91],
    'Science_Score': [88, 85, 90, 87, 92, 84, 89],
    'English_Score': [90, 88, 85, 92, 87, 89, 91]
})

print("="*70)
print("ORIGINAL DATASET")
print("="*70)
print(df)
```

```python
# 1.1 Basic Ranking
print("="*70)
print(" RANKING IN PANDAS")
print("="*70)

print("\n Original Data:")
print(df[['Student', 'Math_Score']])

# Add rank column
df['Math_Rank'] = df['Math_Score'].rank()
print("\n Basic Ranking (default method='average'):")
print(df[['Student', 'Math_Score', 'Math_Rank']])
```

```python
# 1.2 Different Ranking Methods
print("="*70)
print(" DIFFERENT RANKING METHODS")
print("="*70)

# Create data with duplicates
rank_data = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank'],
    'Score': [85, 92, 85, 92, 88, 85]
})

print("\n Data with Duplicates:")
print(rank_data)
```

```python
# Method 1: Average rank (default)
print("\n Method 1: 'average' - Average of ranks")
rank_data['Rank_Avg'] = rank_data['Score'].rank(method='average')
print(rank_data[['Student', 'Score', 'Rank_Avg']])
```

```python
# Method 2: Min rank
print("\n Method 2: 'min' - Lowest rank")
rank_data['Rank_Min'] = rank_data['Score'].rank(method='min')
print(rank_data[['Student', 'Score', 'Rank_Min']])
```

```python
# Method 3: Max rank
print("\n Method 3: 'max' - Highest rank")
rank_data['Rank_Max'] = rank_data['Score'].rank(method='max')
print(rank_data[['Student', 'Score', 'Rank_Max']])
```

```python
# Method 4: First rank (by order in dataset)
print("\n Method 4: 'first' - Order of appearance")
rank_data['Rank_First'] = rank_data['Score'].rank(method='first')
print(rank_data[['Student', 'Score', 'Rank_First']])
```

```python
# Method 5: Dense rank (no gaps)
print("\n Method 5: 'dense' - No gaps in ranks")
rank_data['Rank_Dense'] = rank_data['Score'].rank(method='dense')
print(rank_data[['Student', 'Score', 'Rank_Dense']])
```

```python
# 1.3 Ranking in Descending Order
print("\n Descending Ranking (highest score gets rank 1):")
rank_data['Rank_Desc'] = rank_data['Score'].rank(ascending=False)
print(rank_data[['Student', 'Score', 'Rank_Desc']])
```

```python
# 1.4 Ranking with Percentage (pct=True)
print("\n Percentage Ranking (0-1 scale):")
rank_data['Rank_Percent'] = rank_data['Score'].rank(pct=True)
print(rank_data[['Student', 'Score', 'Rank_Percent']])
print("\nInterpretation: Top 20% = 0.8-1.0")
```
