```python
# Pivot Tables
```

```python
import pandas as pd

df = pd.read_json("students2.json")
print(df)
```

```python
# What is a Pivot Table?
# Pivot table reorganizes data to show summary in a matrix form.
```

```python
# Example idea

# Rows → city
# Columns → subject
# Values → marks
```

```python
# Basic Pivot Table

pd.pivot_table(df, values="marks", index="city", columns="subject")
```

```python
# ndex → rows
# columns → column headers
# values → data to aggregate
```

```python
# Default Aggregation
# By default:

aggfunc = "mean"
```

```python
# Aggregation Function:Sum
pd.pivot_table(df, values="marks", index="city", columns="subject", aggfunc="sum")
```

```python
# Aggregation Function:MAX
pd.pivot_table(df, values="marks", index="city", columns="subject", aggfunc="max")
```

```python
# Aggregation Function:MIN
pd.pivot_table(df, values="marks", index="city", columns="subject", aggfunc="min")
```

```python
# Aggregation Function:COUNT
pd.pivot_table(df, values="marks", index="city", columns="subject", aggfunc="count")
```

```python
# Multiple Aggregations
pd.pivot_table(
    df,
    values="marks",
    index="city",
    columns="subject",
    aggfunc=["mean", "max"]
)
```

```python
import pandas as pd

df = pd.read_json("students2.json")
print(df)
```

```python
# Fill Missing Values in Pivot
vc = pd.pivot_table(
    df,
    values="marks",
    index="city",
    columns="subject",
    fill_value=0
)
# Replaces NaN with 0.
```

```python
pd.pivot_table(vc, values="marks", index="city", columns="subject")
```

```python
# Add Margins (Totals)
pd.pivot_table(
    df,
    values="marks",
    index="city",
    columns="subject",
    margins=True
)
# All → total
```
