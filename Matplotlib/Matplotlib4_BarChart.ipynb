```python
# 2. What is a Bar Chart?
# A Bar Chart represents data using rectangular bars, where:
# -Height (or length) ∝ value
# -Categories are discrete
# -Bars are separated (not continuous)
```

```python
# 3. When to Use Bar Charts?
#  Use bar charts when:
# -Comparing categories
# -Showing discrete values
# -Ranking entities
# -Showing frequency/count
```

```python
import matplotlib.pyplot as plt

categories = ["CSE","IT","ECE","ME"]
values = [120,90,80,70]

plt.bar(categories, values)
plt.show()
```

```python
# Professional Example
import matplotlib.pyplot as plt

branches = ["CSE","IT","ECE","ME"]
students = [120,90,80,70]

plt.bar(branches, students, color='blue')

plt.title("Branch-wise Student Count")
plt.xlabel("Branch")
plt.ylabel("Number of Students")

# plt.show()
```

```python
# Output Interpretation
# Students should observe:
# -CSE has highest enrollment
# -ME has lowest
# -Clear ranking of categories
```

```python
# Customizing Bars
plt.bar(branches, students,
        color='green',
        edgecolor='black',
        width=0.5)
plt.title("Data Example")
```

```python
# | Parameter | Meaning       |
# | --------- | ------------- |
# | color     | fill color    |
# | edgecolor | border        |
# | width     | bar thickness |
```

```python
# Adding Labels on Bars
import matplotlib.pyplot as plt

branches = ["CSE","IT","ECE","ME"]
students = [120,90,80,70]

plt.bar(branches, students)

for i, value in enumerate(students):
    plt.text(i, value+2, str(value), ha='center')

plt.show()
```

```python
# Why Annotation is Important?
# -Improves readability
# -Useful in presentations
# -Required in research papers
```

```python
# # Grouped Bar Chart (Very Important)
# # Used when comparing multiple categories.
# | Branch | Boys | Girls |
# | ------ | ---- | ----- |
# | CSE    | 80   | 40    |
# | IT     | 50   | 40    |
# | ECE    | 45   | 35    |
```

```python
import numpy as np
import matplotlib.pyplot as plt

branches = ["CSE","IT","ECE"]

boys = [80,50,45]
girls = [40,40,35]

x = np.arange(len(branches))
width = 0.35

plt.bar(x - width/2, boys, width, label='Boys')
plt.bar(x + width/2, girls, width, label='Girls')

plt.xticks(x, branches)
plt.legend()
# Without arange(), Matplotlib wouldn’t know where to place the bars. By default, bar charts need numerical positions, not strings. Later, you map these positions back to labels using plt.
# xticks(x, branches) so that instead of showing 0,1,2, the chart displays "CSE","IT","ECE".

plt.title("Gender Distribution")
plt.show()
```

```python
# Interpretation
# -CSE has highest enrollment
# -Gender ratio visible
# -Comparative insight across branches
```

```python
# Stacked Bar Chart
# Used for showing composition within category
branches = ["CSE","IT","ECE"]

boys = [80,50,45]
girls = [40,40,35]

plt.bar(branches, boys, bottom=girls, label='Boys')
plt.bar(branches, girls, label='Girls')

plt.legend()
plt.title("Stacked Bar Chart")
plt.show()
```

```python
# Interpretation
# -Total height = total students
# -Colors show internal composition
```

```python
# Bar Chart for Machine Learning
models = ["SVM","Random Forest","KNN","Neural Net"]
accuracy = [85,90,80,92]

plt.bar(models, accuracy, color='orange')
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.xlabel("Models")
plt.show()
```

```python
# Interpretation:
# 1.Neural Net performs best
# 2.KNN lowest accuracy
# 3.Useful for model selection
```

```python
# Advanced Customization
plt.figure(figsize=(8,5))

plt.bar(models, accuracy,
        color=['blue','orange','yellow','black'])

plt.grid(axis='y', linestyle='-.', alpha=0.8)
# alpha(0.9):gives line darkness
```
