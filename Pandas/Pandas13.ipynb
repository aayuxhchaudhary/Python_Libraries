```python
# Two main formats:Wide and Long
# Wide format: 
# city   Math  Science
# Delhi   90     85
```

```python
# Long format
# city  subject  marks
# Delhi Math      90
# Delhi Science   85

# NOTE: Wide = more columns
#       Long = more rows
```

```python
import pandas as pd

df = pd.read_json("students_wide.json")
print(df)
```

```python
# melt() (Wide → Long)
# melt() = columns ko rows me convert karna
df_melt = pd.melt(
    df,
    id_vars=["name"],
    var_name="subject",
    value_name="marks"
)

print(df_melt)
# id_vars → columns to keep fixed
# others become rows
```

```python
import pandas as pd

df = pd.read_json("students_long.json")
print(df)
```

```python
# df.pivot(index, columns, values)
```

```python
# pivot() (Long → Wide)
df_pivot = df_melt.pivot(
    index="name",
    columns="subject",
    values="marks"
)
print(df_pivot)
# pivot() fails if duplicate values exist
```

```python
# pivot() (Long → Wide)
df_pivot = df_melt.pivot_table(
    index="name",
    columns="subject",
    values="marks"
)
print(df_pivot)
# pivot() fails if duplicate values exist
```

```python
# pivot_table():Because real data has duplicates.
df_pivot = pd.pivot_table(
    df_melt,
    values="marks",
    index="name",
    columns="subject",
    aggfunc="mean"
)
print(df_pivot)
```

```python
# stack() and unstack()
df_stack = df.set_index("name").stack()
print(df_stack)
```

```python
# unstack() → rows → columns
vc = df_stack.unstack()
print(vc)
```

```python
import pandas as pd

xc = pd.read_json("students_long.json")
print(xc)
```

```python
df_stack2 = xc.set_index("name").stack()
print(df_stack2)
```

```python
# unstack() → rows → columns
oc = df_stack2.unstack()
print(oc)
```
