```python
# PART 3: RESAMPLING TIME SERIES DATA

# 3.1 What is Resampling?
# Resampling changes the frequency of your time series data (e.g., daily to monthly, hourly to daily).

print("="*70)
print(" RESAMPLING TIME SERIES DATA")
print("="*70)

# Create daily data for 3 months
daily_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=90, freq='D'),
    'Sales': np.random.randint(1000, 5000, 90),
    'Customers': np.random.randint(50, 200, 90)
})
daily_data.set_index('Date', inplace=True)

print(" Daily Data Sample (First 10 days):")
print(daily_data.head(10))
print(f"\nTotal daily records: {len(daily_data)}")
```

```python
# 1. Downsampling (higher to lower frequency) - SUMMARIZING
print("\n" + "="*50)
print(" DOWNSAMPLING: Daily → Weekly")
print("="*50)

print("\n Weekly Total Sales:")
weekly_sum = daily_data.resample('W').sum()
print(weekly_sum.head(8))
print(f"\nWeekly records: {len(weekly_sum)} (reduced from {len(daily_data)})")

print("\n Weekly Average Sales:")
weekly_mean = daily_data.resample('W').mean()
print(weekly_mean.head(8))

print("\n Weekly Statistics (Multiple Aggregations):")
weekly_stats = daily_data.resample('W').agg({
    'Sales': ['sum', 'mean', 'min', 'max', 'std'],
    'Customers': ['sum', 'mean']
})
print(weekly_stats.head(5))
```

```python
# 2. Downsampling to Monthly
print("\n" + "="*50)
print(" DOWNSAMPLING: Daily → Monthly")
print("="*50)

monthly_data = daily_data.resample('M').agg({
    'Sales': ['sum', 'mean', 'min', 'max'],
    'Customers': ['sum', 'mean']
})
print(monthly_data)
print(f"\nMonthly records: {len(monthly_data)}")
```

```python
# 3. Upsampling (lower to higher frequency) - FILLING IN GAPS
print("\n" + "="*50)
print(" UPSAMPLING: Weekly → Daily")
print("="*50)

# First create weekly data
weekly_data = daily_data.resample('W').mean()
print("Original Weekly Data:")
print(weekly_data.head(5))

print("\n Upsampled to Daily (Forward Fill):")
daily_ffill = weekly_data.resample('D').ffill()  # Forward fill
print(daily_ffill.head(15))

print("\n Upsampled to Daily (Interpolation):")
daily_interp = weekly_data.resample('D').interpolate()  # Linear interpolation
print(daily_interp.head(15))
```

```python
# 4. Different Resampling Frequencies
print("\n" + "="*50)
print(" COMMON RESAMPLING FREQUENCIES")
print("="*50)

freq_demo = pd.DataFrame({
    'Code': ['D', 'W', 'M', 'Q', 'Y', 'B', 'H'],
    'Meaning': ['Daily', 'Weekly', 'Month End', 'Quarter End', 'Year End', 'Business Day', 'Hourly'],
    'Use Case': ['Daily sales', 'Weekly reports', 'Monthly summary', 'Quarterly results', 'Annual report', 'Stock market', 'Website traffic']
})
print(freq_demo.to_string(index=False))
```

```python
# 5. Practical Example: Finding Busiest Week
print("\n" + "="*50)
print(" PRACTICAL: Finding Best Week for Sales")
print("="*50)

weekly_sales = daily_data.resample('W')['Sales'].sum()
print("Weekly Sales Totals:")
print(weekly_sales)

print(f"\n Best Week: {weekly_sales.idxmax().strftime('%Y-%m-%d')}")
print(f" Sales: ${weekly_sales.max():,.2f}")

print(f"\n Worst Week: {weekly_sales.idxmin().strftime('%Y-%m-%d')}")
print(f" Sales: ${weekly_sales.min():,.2f}")
```

```python
# PART 4: ROLLING WINDOW FUNCTIONS

# 4.1 What are Rolling Windows?
# Rolling windows calculate statistics over a moving window of data points. Think of it as looking at the last N days to understand current trends.

print("="*70)
print("ROLLING WINDOW FUNCTIONS (Moving Averages)")
print("="*70)

# Create sample data with some noise
np.random.seed(42)
rolling_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=30, freq='D'),
    'Daily_Sales': [100, 105, 102, 108, 115, 112, 118, 125, 122, 128,
                    135, 132, 138, 145, 142, 148, 155, 152, 158, 165,
                    162, 168, 175, 172, 178, 185, 182, 188, 195, 192]
})
rolling_data.set_index('Date', inplace=True)

print(" Original Daily Sales Data:")
print(rolling_data.head(10))
```

```python
# 1. Simple Moving Average (SMA)
print("\n" + "="*50)
print("MOVING AVERAGE - Smoothing the Data")
print("="*50)

print("\n 7-Day Moving Average (Last week's average):")
rolling_data['MA_7'] = rolling_data['Daily_Sales'].rolling(window=7).mean()
print(rolling_data[['Daily_Sales', 'MA_7']].head(15))

print("\n 14-Day Moving Average (Last two weeks average):")
rolling_data['MA_14'] = rolling_data['Daily_Sales'].rolling(window=14).mean()
print(rolling_data[['Daily_Sales', 'MA_7', 'MA_14']].head(20))
```

```python
# Explanation of moving average
print("\n UNDERSTANDING MOVING AVERAGES:")
print("Day 7 MA = Average of Days 1-7")
print("Day 8 MA = Average of Days 2-8")
print("Day 9 MA = Average of Days 3-9")
print("This 'smooths out' daily fluctuations to show the true trend!")
```

```python
# 2. Rolling Sum
print("\n" + "="*50)
print(" ROLLING SUM - Weekly Totals")
print("="*50)

rolling_data['Week_Total'] = rolling_data['Daily_Sales'].rolling(window=7).sum()
print(rolling_data[['Daily_Sales', 'Week_Total']].head(15))

print("\n INTERPRETATION:")
print("Day 7 Week_Total = Sum of Days 1-7")
print("Day 8 Week_Total = Sum of Days 2-8")
print("Day 9 Week_Total = Sum of Days 3-9")
```

```python
# 3. Rolling Statistics
print("\n" + "="*50)
print(" ROLLING STATISTICS")
print("="*50)

rolling_data['Rolling_Std'] = rolling_data['Daily_Sales'].rolling(window=7).std()
rolling_data['Rolling_Max'] = rolling_data['Daily_Sales'].rolling(window=7).max()
rolling_data['Rolling_Min'] = rolling_data['Daily_Sales'].rolling(window=7).min()

print("7-Day Rolling Statistics:")
print(rolling_data[['Daily_Sales', 'Rolling_Max', 'Rolling_Min', 'Rolling_Std']].head(15))

print("\n WHAT THESE MEAN:")
print("• Rolling_Max: Highest sales in last 7 days")
print("• Rolling_Min: Lowest sales in last 7 days")
print("• Rolling_Std: How much sales vary day-to-day")
```

```python
# 4. Rolling with min_periods (handling beginning)
print("\n" + "="*50)
print(" HANDLING STARTING PERIODS (min_periods)")
print("="*50)

rolling_data['MA_With_Min'] = rolling_data['Daily_Sales'].rolling(window=7, min_periods=3).mean()
print(rolling_data[['Daily_Sales', 'MA_7', 'MA_With_Min']].head(10))

print("\n min_periods=3 means:")
print("• First 2 days: Not enough data → NaN")
print("• Day 3: Average of days 1-3 (only 3 values)")
print("• Day 4: Average of days 1-4")
print("• Day 7+: Average of full 7 days")
```

```python
# 5. Practical Business Application
print("\n" + "="*50)
print(" PRACTICAL: Identifying Sales Trends")
print("="*50)

# Add more data for trend analysis
trend_data = rolling_data.copy()
trend_data['MA_7'] = trend_data['Daily_Sales'].rolling(7).mean()
trend_data['MA_14'] = trend_data['Daily_Sales'].rolling(14).mean()

print("\n Trend Direction Detection:")
trend_data['Trend'] = trend_data['MA_7'] - trend_data['MA_14']
print(trend_data[['Daily_Sales', 'MA_7', 'MA_14', 'Trend']].tail(15))

print("\n INTERPRETING TRENDS:")
print("• Positive Trend: 7-day MA > 14-day MA (Sales increasing)")
print("• Negative Trend: 7-day MA < 14-day MA (Sales decreasing)")
print("• Neutral Trend: 7-day MA ≈ 14-day MA (Sales stable)")
```

```python
# PART 5: EXPANDING WINDOW FUNCTIONS
# 5.1 What are Expanding Windows?
# Expanding windows include ALL previous data points - perfect for cumulative calculations.

print("="*70)
print(" EXPANDING WINDOW FUNCTIONS (Cumulative Calculations)")
print("="*70)

# Create sample data
expand_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=20, freq='D'),
    'Daily_Sales': [100, 110, 105, 120, 115, 130, 125, 140, 135, 150,
                    145, 160, 155, 170, 165, 180, 175, 190, 185, 200]
})
expand_data.set_index('Date', inplace=True)

print(" Daily Sales Data:")
print(expand_data)
```

```python
# 1. Cumulative Sum (Running Total)
print("\n" + "="*50)
print(" CUMULATIVE SUM - Year-to-Date (YTD) Sales")
print("="*50)

expand_data['Cumulative_Sales'] = expand_data['Daily_Sales'].expanding().sum()
print(expand_data)

print("\n INTERPRETATION:")
print("Day 1: Total = Day 1")
print("Day 2: Total = Day 1 + Day 2")
print("Day 3: Total = Day 1 + Day 2 + Day 3")
print("This answers: 'How much have we sold so far this year?'")
```

```python
# 2. Cumulative Average (Running Average)
print("\n" + "="*50)
print(" CUMULATIVE AVERAGE - Overall Performance")
print("="*50)

expand_data['Cumulative_Avg'] = expand_data['Daily_Sales'].expanding().mean()
print(expand_data[['Daily_Sales', 'Cumulative_Sales', 'Cumulative_Avg']])

print("\n INTERPRETATION:")
print("Day 1 Avg = Day 1")
print("Day 2 Avg = (Day 1 + Day 2) / 2")
print("Day 3 Avg = (Day 1 + Day 2 + Day 3) / 3")
print("This shows how your overall average changes over time")
```

```python

# 3. Cumulative Maximum and Minimum
print("\n" + "="*50)
print(" CUMULATIVE MAX/MIN - Records")
print("="*50)

expand_data['Record_High'] = expand_data['Daily_Sales'].expanding().max()
expand_data['Record_Low'] = expand_data['Daily_Sales'].expanding().min()
expand_data['Is_Record_High'] = expand_data['Daily_Sales'] == expand_data['Record_High']
expand_data['Is_Record_Low'] = expand_data['Daily_Sales'] == expand_data['Record_Low']

print(expand_data[['Daily_Sales', 'Record_High', 'Is_Record_High', 'Record_Low', 'Is_Record_Low']])

print(f"\n Number of record highs: {expand_data['Is_Record_High'].sum()}")
print(f" Number of record lows: {expand_data['Is_Record_Low'].sum()}")
```

```python
# 4. Cumulative Standard Deviation
print("\n" + "="*50)
print("CUMULATIVE STANDARD DEVIATION - Consistency")
print("="*50)

expand_data['Cumulative_Std'] = expand_data['Daily_Sales'].expanding().std()
print(expand_data[['Daily_Sales', 'Cumulative_Avg', 'Cumulative_Std']])

print("\n INTERPRETATION:")
print("• Low Std = Consistent sales")
print("• High Std = Volatile sales")
```

```python
# 5. Real-World Business Applications
print("\n" + "="*50)
print(" REAL-WORLD APPLICATIONS")
print("="*50)

print("\n Application 1: Year-to-Date (YTD) Performance:")
print(f"Total YTD Sales: ${expand_data['Cumulative_Sales'].iloc[-1]:,.2f}")
print(f"Average Daily Sales: ${expand_data['Cumulative_Avg'].iloc[-1]:.2f}")
print(f"Best Day So Far: ${expand_data['Record_High'].iloc[-1]:,.2f}")
```

```python
print("\n Application 2: Monthly Target Tracking:")
daily_target = 150
expand_data['Daily_Target_Met'] = expand_data['Daily_Sales'] >= daily_target
expand_data['Target_Success_Rate'] = (expand_data['Daily_Target_Met'].expanding().sum() / 
                                       expand_data['Daily_Target_Met'].expanding().count()) * 100
print(expand_data[['Daily_Sales', 'Daily_Target_Met', 'Target_Success_Rate']].tail(10))

print(f"\n Overall Target Success Rate: {expand_data['Target_Success_Rate'].iloc[-1]:.1f}%")
```

```python
print("\n Application 3: Running Total with Goal:")
yearly_goal = 50000
expand_data['Remaining_to_Goal'] = yearly_goal - expand_data['Cumulative_Sales']
expand_data['Percent_Complete'] = (expand_data['Cumulative_Sales'] / yearly_goal) * 100
print(expand_data[['Daily_Sales', 'Cumulative_Sales', 'Percent_Complete', 'Remaining_to_Goal']].tail(10))
```

```python
# PART 6: EXPONENTIAL WEIGHTED FUNCTIONS
# 6.1 What is Exponential Weighting?
# Gives MORE weight to RECENT observations - great for forecasting and trend detection.

print("="*70)
print(" EXPONENTIAL WEIGHTED FUNCTIONS (Focus on Recent Data)")
print("="*70)

# Create sample data
np.random.seed(42)
ewm_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=30, freq='D'),
    'Sales': [100, 102, 101, 105, 108, 107, 200, 110, 112, 115,
              118, 117, 120, 122, 121, 125, 128, 127, 130, 132,
              135, 134, 138, 140, 137, 300, 142, 145, 148, 150]
})
ewm_data.set_index('Date', inplace=True)

print(" Sales Data with Spikes:")
print(ewm_data.head(15))
```

```python
# 1. Exponential Weighted Moving Average (EWMA)
print("\n" + "="*50)
print("EXPONENTIAL WEIGHTED MOVING AVERAGE")
print("="*50)

ewm_data['EWMA_7'] = ewm_data['Sales'].ewm(span=7).mean()
print(ewm_data[['Sales', 'EWMA_7']].head(15))

print("\n HOW EWMA IS DIFFERENT:")
print("• Simple MA: All 7 days have EQUAL weight")
print("• EWMA: Recent days have MORE weight")
print("• EWMA reacts faster to recent changes!")
```

```python
# 2. Compare Simple MA vs EWMA
print("\n" + "="*50)
print(" COMPARISON: Simple MA vs EWMA")
print("="*50)

ewm_data['SMA_7'] = ewm_data['Sales'].rolling(7).mean()
print(ewm_data[['Sales', 'SMA_7', 'EWMA_7']].head(20))

print("\n KEY DIFFERENCES OBSERVED:")
print("• SMA: Changes slowly, lags behind")
print("• EWMA: Responds quickly to new trends")
print("• EWMA better for detecting recent patterns")
```

```python
# 3. Different Span Values
print("\n" + "="*50)
print(" DIFFERENT SPAN VALUES")
print("="*50)

ewm_data['EWMA_14'] = ewm_data['Sales'].ewm(span=14).mean()
ewm_data['EWMA_30'] = ewm_data['Sales'].ewm(span=30).mean()

print(ewm_data[['Sales', 'EWMA_7', 'EWMA_14', 'EWMA_30']].head(20))

print("\ CHOOSING SPAN VALUE:")
print("• Small span (7): Reacts quickly, more sensitive")
print("• Medium span (14): Balanced approach")
print("• Large span (30): Very smooth, less sensitive")
```

```python
# 4. Exponential Weighted Standard Deviation
print("\n" + "="*50)
print(" EXPONENTIAL WEIGHTED VOLATILITY")
print("="*50)

ewm_data['EWM_Volatility'] = ewm_data['Sales'].ewm(span=7).std()
print(ewm_data[['Sales', 'EWMA_7', 'EWM_Volatility']].head(20))

print("\n USE CASE:")
print("• High volatility: Sales are unstable")
print("• Low volatility: Sales are consistent")
```

```python
# PART 7: SHIFTING AND LAGGING
# 7.1 What is Shifting?
# Moving data forward or backward in time to compare with previous periods.


print("="*70)
print("SHIFTING AND LAGGING (Comparing Time Periods)")
print("="*70)

# Create sample data
shift_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=15, freq='D'),
    'Sales': [100, 110, 105, 120, 115, 130, 125, 140, 135, 150, 145, 160, 155, 170, 165]
})
shift_data.set_index('Date', inplace=True)

print("Original Sales Data:")
print(shift_data)
```

```python
# 1. Shift Forward (Lag) - Compare with previous day
print("\n" + "="*50)
print(" SHIFT FORWARD - Previous Day Comparison")
print("="*50)

shift_data['Previous_Day'] = shift_data['Sales'].shift(1)
shift_data['Daily_Change'] = shift_data['Sales'] - shift_data['Previous_Day']
shift_data['Daily_Pct_Change'] = (shift_data['Daily_Change'] / shift_data['Previous_Day']) * 100

print(shift_data[['Sales', 'Previous_Day', 'Daily_Change', 'Daily_Pct_Change']])

print("\n BUSINESS INSIGHTS:")
print(f"• Days with sales increase: {(shift_data['Daily_Change'] > 0).sum()}")
print(f"• Days with sales decrease: {(shift_data['Daily_Change'] < 0).sum()}")
print(f"• Average daily change: ${shift_data['Daily_Change'].mean():.2f}")
```

```python
# 2. Shift Backward (Lead) - Look ahead
print("\n" + "="*50)
print(" SHIFT BACKWARD - Next Day Prediction")
print("="*50)

shift_data['Next_Day'] = shift_data['Sales'].shift(-1)
print(shift_data[['Sales', 'Next_Day']].head(10))

print("\n USE CASE: Planning inventory for tomorrow")
```

```python
# 3. Multiple Period Shifts
print("\n" + "="*50)
print(" WEEK-OVER-WEEK COMPARISON (7-day shift)")
print("="*50)

shift_data['Sales_7_Days_Ago'] = shift_data['Sales'].shift(7)
shift_data['WoW_Change'] = shift_data['Sales'] - shift_data['Sales_7_Days_Ago']
shift_data['WoW_Pct_Change'] = (shift_data['WoW_Change'] / shift_data['Sales_7_Days_Ago']) * 100

print(shift_data[['Sales', 'Sales_7_Days_Ago', 'WoW_Change', 'WoW_Pct_Change']])

print("\n WEEK-OVER-WEEK ANALYSIS:")
print(f"• Average WoW growth: {shift_data['WoW_Pct_Change'].mean():.1f}%")
print(f"• Best WoW growth: {shift_data['WoW_Pct_Change'].max():.1f}%")
print(f"• Worst WoW decline: {shift_data['WoW_Pct_Change'].min():.1f}%")
```

```python
# 4. Year-over-Year Comparison (Monthly data example)
print("\n" + "="*50)
print(" YEAR-OVER-YEAR COMPARISON")
print("="*50)

yearly_data = pd.DataFrame({
    'Date': pd.date_range('2023-01-01', periods=24, freq='M'),
    'Sales': [100, 110, 105, 120, 115, 130, 125, 140, 135, 150, 145, 160,
              155, 170, 165, 180, 175, 190, 185, 200, 195, 210, 205, 220]
})
yearly_data.set_index('Date', inplace=True)

yearly_data['Sales_Prev_Year'] = yearly_data['Sales'].shift(12)
yearly_data['YoY_Growth'] = ((yearly_data['Sales'] - yearly_data['Sales_Prev_Year']) / 
                              yearly_data['Sales_Prev_Year']) * 100

print(yearly_data.tail(15))
print(f"\n Year-over-Year Growth Rate: {yearly_data['YoY_Growth'].iloc[-1]:.1f}%")
```

```python
# 5. Practical: Sales Momentum
print("\n" + "="*50)
print(" SALES MOMENTUM INDICATOR")
print("="*50)

shift_data['Momentum'] = shift_data['Daily_Change'].rolling(3).mean()
shift_data['Strong_Momentum'] = shift_data['Momentum'] > 10
shift_data['Weak_Momentum'] = shift_data['Momentum'] < -10

print(shift_data[['Sales', 'Daily_Change', 'Momentum', 'Strong_Momentum', 'Weak_Momentum']].tail(10))

print("\n MOMENTUM INTERPRETATION:")
print("• Strong momentum: Sales consistently increasing")
print("• Weak momentum: Sales consistently decreasing")
print("• Use for identifying trends early")
```

```python
# PART 8: DIFFERENCE OPERATIONS
# 8.1 Calculating Changes Between Periods

print("="*70)
print(" DIFFERENCE OPERATIONS (Measuring Change)")
print("="*70)

# Create sample data
diff_data = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=20, freq='D'),
    'Stock_Price': [100, 102, 101, 105, 108, 107, 110, 112, 111, 115,
                    118, 117, 120, 122, 121, 125, 128, 127, 130, 132]
})
diff_data.set_index('Date', inplace=True)

print(" Stock Price Data:")
print(diff_data)
```

```python
# 1. First Difference (Period-to-Period Change)
print("\n" + "="*50)
print(" FIRST DIFFERENCE - Daily Change")
print("="*50)

diff_data['Daily_Change'] = diff_data['Stock_Price'].diff()
diff_data['Daily_Return'] = diff_data['Stock_Price'].pct_change() * 100

print(diff_data[['Stock_Price', 'Daily_Change', 'Daily_Return']])

print("\n STATISTICS:")
print(f"• Average daily change: ${diff_data['Daily_Change'].mean():.2f}")
print(f"• Average daily return: {diff_data['Daily_Return'].mean():.2f}%")
print(f"• Best day: +${diff_data['Daily_Change'].max():.2f}")
print(f"• Worst day: ${diff_data['Daily_Change'].min():.2f}")
```

```python

# 2. Multiple Period Difference
print("\n" + "="*50)
print(" WEEK-OVER-WEEK CHANGE (7-day difference)")
print("="*50)

diff_data['Weekly_Change'] = diff_data['Stock_Price'].diff(7)
diff_data['Weekly_Return'] = diff_data['Stock_Price'].pct_change(7) * 100

print(diff_data[['Stock_Price', 'Weekly_Change', 'Weekly_Return']])

print(f"\n Weekly Performance:")
print(f"• Average weekly change: ${diff_data['Weekly_Change'].mean():.2f}")
print(f"• Average weekly return: {diff_data['Weekly_Return'].mean():.2f}%")
```

```python
# 3. Second Difference (Change of Change - Acceleration)
print("\n" + "="*50)
print(" SECOND DIFFERENCE - Acceleration/Deceleration")
print("="*50)

diff_data['Daily_Change'] = diff_data['Stock_Price'].diff()
diff_data['Acceleration'] = diff_data['Daily_Change'].diff()

print(diff_data[['Stock_Price', 'Daily_Change', 'Acceleration']])

print("\n INTERPRETATION:")
print("• Positive Acceleration: Price increasing faster")
print("• Negative Acceleration: Price increasing slower or decreasing")
print("• Zero Acceleration: Steady growth")
```

```python
# 4. Log Returns (Better for financial analysis)
print("\n" + "="*50)
print(" LOG RETURNS (For Financial Data)")
print("="*50)

diff_data['Log_Return'] = np.log(diff_data['Stock_Price'] / diff_data['Stock_Price'].shift(1)) * 100

print(diff_data[['Stock_Price', 'Daily_Return', 'Log_Return']])

print("\n WHY LOG RETURNS?")
print("• Log returns are additive over time")
print("• Better for statistical analysis")
print("• Symmetric for gains and losses")
```

```python
# 5. Identifying Trend Changes
print("\n" + "="*50)
print(" IDENTIFYING TREND CHANGES")
print("="*50)

# Calculate 5-day average change
diff_data['Avg_5Day_Change'] = diff_data['Daily_Change'].rolling(5).mean()
diff_data['Trend_Direction'] = diff_data['Avg_5Day_Change'].apply(
    lambda x: 'Up' if x > 0 else 'Down' if x < 0 else 'Flat'
)
diff_data['Trend_Change'] = diff_data['Trend_Direction'] != diff_data['Trend_Direction'].shift(1)

print(diff_data[['Stock_Price', 'Avg_5Day_Change', 'Trend_Direction', 'Trend_Change']])

trend_changes = diff_data[diff_data['Trend_Change']]
print(f"\n Number of trend changes: {len(trend_changes)}")
```
