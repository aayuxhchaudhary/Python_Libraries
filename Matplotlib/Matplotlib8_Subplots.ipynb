```python
# 4. Methods to Create Subplots
# 1. plt.subplot() → Basic (index-based)
# 2. plt.subplots() → Advanced (Object-Oriented, recommended)
```

```python
# Using plt.subplot()
 plt.subplot(rows, cols, index)

# Parameters:
# rows → number of rows in grid
# cols → number of columns
# index → position of the current plot (starts from 1)
```

```python
import matplotlib.pyplot as plt

# -------- Plot 1 --------
plt.subplot(2,2,1)  # row=2, col=2, position=1
plt.plot([1,2,3],[1,4,9])
plt.title("Line Plot")

# -------- Plot 2 --------
plt.subplot(2,2,2)
plt.bar([1,2,3],[3,2,5])
plt.title("Bar Chart")

# -------- Plot 3 --------
plt.subplot(2,2,3)
plt.scatter([1,2,3],[5,3,6])
plt.title("Scatter Plot")

# -------- Plot 4 --------
plt.subplot(2,2,4)
plt.hist([1,2,2,3,3,3,4])
plt.title("Histogram")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show all plots
plt.show()
```

```python
# Limitations of subplot()
# -Less flexible
# -Hard for complex layouts
# Prefer plt.subplots()
```

```python
# Using plt.subplots()
# fig, ax = plt.subplots(2,2)
# A 2×2 grid of plots means that your figure is divided into 4 smaller plotting areas arranged like a table with 2 rows and 2 columns:
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,2)

ax[0,0].plot([1,2,3],[1,4,9])
ax[0,0].set_title("Quadratic")
ax[0,0].set_xlabel("X-axis")
ax[0,0].set_ylabel("Y-axis")
ax[0,0].legend(['Data'])

ax[0,1].plot([1,2,3],[1,2,3])
ax[0,1].set_title("Linear")
ax[0,1].set_xlabel("X-axis")
ax[0,1].set_ylabel("Y-axis")
ax[0,1].legend(['Data'])

ax[1,0].bar([1,2,3],[3,2,5])
ax[1,0].set_title("Bar")
ax[1,0].set_xlabel("Categories")
ax[1,0].set_ylabel("Values")
ax[1,0].legend(['Data'])


ax[1,1].hist([1,2,2,3,3,3])
ax[1,1].set_title("Histogram")
ax[1,1].set_xlabel("Bins")
ax[1,1].set_ylabel("Count")
ax[1,1].legend(['Data'])

plt.show()

# ax is a 2D NumPy array of axes objects, so we can access each subplot with ax[row, col].
# Creates:
# [ ax[0,0]   ax[0,1] ]
# [ ax[1,0]   ax[1,1] ]
```

```python
# | Component | Meaning         |
# | --------- | --------------- |
# | Figure    | Entire canvas   |
# | Axes      | Individual plot |
# | Axis      | X/Y scale       |
```

```python
# Single Row Subplots
fig, ax = plt.subplots(1,3) #1×3 grid of plots — meaning one row with three columns of subplots, all side by side in a single figure.

ax[0].plot([1,2,3],[2,4,6]) #Plots points (1,2), (2,4), (3,6).
ax[1].plot([1,2,3],[3,6,9]) #Plots points (1,3), (2,6), (3,9).
ax[2].plot([1,2,3],[4,8,12]) #Plots points (1,4), (2,8), (3,12).

plt.show()
```

```python
# Single Column Subplots
fig, ax = plt.subplots(3,1) #Makes a figure with 3 rows and 1 column of subplots.

ax[0].plot([1,2,3],[2,4,6]) #Plots points (1,2), (2,4), (3,6).
ax[1].plot([1,2,3],[3,6,9]) #Plots points (1,3), (2,6), (3,9)
ax[2].plot([1,2,3],[4,8,12]) #Plots points (1,4), (2,8), (3,12)

plt.show()
```

```python
# Adjusting Spacing

# plt.tight_layout()
# When we create multiple subplots, labels and titles often overlap or get cut off.
# plt.tight_layout() automatically adjusts spacing between subplots so everything fits neatly

# Fixes overlapping titles and labels.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4)) #Creates a new figure with a custom size: 6 inches wide, 4 inches tall

plt.subplot(2,1,1) #divides the figure into 2 rows × 1 column, and selects the first subplot (top).
plt.plot([1,2,3],[1,4,9]) #Plots points (1,1), (2,4), (3,9)
plt.title("This is a very long title for Plot 1")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

plt.subplot(2,1,2) #selects the second subplot (bottom).
plt.plot([1,2,3],[9,4,1]) #Plots points (1,9), (2,4), (3,1)
plt.title("This is a very long title for Plot 2")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

plt.show()
```

```python
# Shared Axes in Subplots (sharex=True)
# When we create multiple subplots, each subplot normally has its own independent axes.
# With shared axes, multiple plots use the same scale, which makes comparison much clearer
```

```python
# Shared Axes (Advanced)
fig, ax = plt.subplots(2,1, sharex=True)
# Breakdown:
# -2,1 → 2 rows, 1 column (vertical layout)
# -ax → array of 2 Axes objects (ax[0], ax[1])
# -sharex=True → both plots use the same X-axis scale

# Benefits:
# -aligned comparison
# -same scale
```

```python
# Example: ML Training vs Validation
import matplotlib.pyplot as plt

epochs = [1,2,3,4,5]

train_loss = [0.9,0.7,0.5,0.3,0.2]
val_loss   = [1.0,0.8,0.6,0.5,0.55]

fig, ax = plt.subplots(2,1, sharex=True)

ax[0].plot(epochs, train_loss)
ax[0].set_title("Training Loss")

ax[1].plot(epochs, val_loss)
ax[1].set_title("Validation Loss")

plt.show()
```

```python
# Shared Axes in Subplots (sharey=True)

# fig, ax = plt.subplots(1,2, sharey=True)

# Breakdown:
# -1,2 → 1 row, 2 columns → two plots placed side by side
# -fig → the Figure object (entire canvas)
# -ax → array of Axes objects → [ax[0], ax[1]]
# -sharey=True → both plots share the same Y-axis scale

# Used when comparing same metric across categories
```

```python
# Example: Comparing Two Models
fig, ax = plt.subplots(1,2, sharey=True)
# fig, ax = plt.subplots(1,2)
x = [1,2,3,4]

model_a = [70,75,80,85]
model_b = [65,70,78,82]

ax[0].plot(x, model_a)
ax[0].set_title("Model A")

ax[1].plot(x, model_b)
ax[1].set_title("Model B")

plt.show()

# Both plots:
# -have independent X-axes
# -share the same Y-axis
```
