```python
# Add New Column
import pandas as pd

df = pd.read_json("student.json")

df["bonus"] = 5

print(df)
# One value assigned → applies to all rows
```

```python
# Add New Column
import pandas as pd

df = pd.read_json("student.json")

df["bonus"] = 5

df["class"] = "MBA"

print(df)
# One value assigned → applies to all rows
```

```python
# Add Column Using Existing Column

import pandas as pd

df = pd.read_json("student.json")

# df["total_marks"] = df["marks"] + 10

print(df)
```

```python
# Add Column Using Existing Column

import pandas as pd

df = pd.read_json("student.json")

df["total_marks"] = df["marks"] - 10

print(df)
```

```python
# Add Percentage Column
import pandas as pd

df = pd.read_json("student.json")

df["total_Marks"] = df["Eng"]+df["Hindi"]+df["Maths"]

df["Average"]=df["total_Marks"]/3

mean=df["total_Marks"].mean()

print(df)
print(mean)
```

```python
# Add Percentage Column
import pandas as pd

df = pd.read_json("student.json")

df["percentage"] = (df["marks"] / 100) * 100
print(df)
```

```python
# Update Existing Column
import pandas as pd

df = pd.read_json("student.json")

df["marks"] = df["marks"] + 5

print(df)
```

```python
# Conditional Update Using loc[]
import pandas as pd

df = pd.read_json("student.json")

# df.loc[df["total_Marks"] < 90, "grade"] = "B"
df.loc[df["Maths"] >= 90, "grade"] = "A"

print(df)
```

```python
# Conditional Update Using loc[]
import pandas as pd

df = pd.read_json("student.json")

# df.loc[df["total_Marks"] < 90, "grade"] = "B"
# df.loc[df["Maths"] >= 90, "grade"] = "A"
df["Maths"] >= 90, "grade" = "A"
print(df)
```

```python
# Rename Column
import pandas as pd

df = pd.read_json("student.json")

df.rename(columns={"Maths":"Mathematics"}, inplace=True)

print(df)
```

```python
# Rename Multiple Columns
import pandas as pd

df = pd.read_json("student.json")

df.rename(columns={
    "name":"student_name",
    "age":"student_age"
}, inplace=True)

print(df)
```

```python
# Delete Column using drop()
import pandas as pd

df = pd.read_json("student.json")

df.drop("age", axis=1, inplace=True)

print(df)

# axis=1 means column
```

```python
# Delete Multiple Columns
import pandas as pd

df = pd.read_json("student.json")

df.drop(["city", "Eng"], axis=1, inplace=True)

print(df)
```

```python
# Delete Row

import pandas as pd

df = pd.read_json("student.json")

df.drop(0, axis=0, inplace=True)

print(df)

# axis=0 means row
```

```python
# Delete Row

import pandas as pd

df = pd.read_json("student.json")

df.drop(0, axis=0, inplace=True)

print(df)

# axis=0 means row
```

```python
# Insert Column at Specific Position
import pandas as pd

df = pd.read_json("student.json")

df.insert(1, "roll_no", [101,102,103,104])

print(df)
# insert(position, column_name, values)
```

```python
# Apply Function to Column
import pandas as pd

df = pd.read_json("student.json")

df["name"] = df["name"].str.upper()

print(df)
```

```python
# Create Column with Condition
import pandas as pd

df = pd.read_json("student.json")

df["result"] = df["Eng"].apply(lambda x: "Pass" if x >= 70 else "Fail")

print(df)
```

```python
# Replace Values
import pandas as pd

df = pd.read_json("student.json")

df["city"] = df["city"].replace("Delhi", "New Delhi")

print(df)
```
