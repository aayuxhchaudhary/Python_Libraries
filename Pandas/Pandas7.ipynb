```python
import pandas as pd

df = pd.read_json("students.json")
print(df)
```

```python
# Selecting Multiple Columns
import pandas as pd

df = pd.read_json("students.json")

print(df[["name","marks"]])
```

```python
# Selecting Rows and Columns Together
import pandas as pd

df = pd.read_json("students.json")

print(df.loc[0:5, ["name", "marks"]])
```

```python
# Filtering Rows
import pandas as pd

df = pd.read_json("students.json")

print(df[df["marks"] >= 95])
```

```python
# Filtering Rows
import pandas as pd

df = pd.read_json("students.json")

bf=(df[df["marks"] > 85])
print(bf)
print("\n")
print(bf[["name","marks"]])
```

```python
# Multiple Conditions
import pandas as pd

df = pd.read_json("students.json")

print(df[(df["marks"] > 85) & (df["age"] > 22)])
```

```python
#We have to use "&" instead of "and" in pandas json to apply mutiple conditions over a statement!!!!!!!
import pandas as pd

df = pd.read_json("students.json")

print(df[(df["marks"] > 85) and (df["age"] > 22)])   #error
```

```python
# Sorting Data
import pandas as pd

df = pd.read_json("students.json")

# Force Pandas to maintain original order during ties
print(df.sort_values("age"))
```

```python
import pandas as pd

df = pd.read_json("students.json")

print(df.sort_values("marks", ascending=False)) #There is no descending function in pandas, We use either True or False
```

```python
import pandas as pd

df = pd.read_json("students.json")

print(df.sort_values("marks", ascending=False))
```
