```python
import pandas as pd
import numpy as np

# Create sample dataset
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace'],
    'Math_Score': [85, 92, 78, 92, 88, 85, 91],
    'Science_Score': [88, 85, 90, 87, 92, 84, 89],
    'English_Score': [90, 88, 85, 92, 87, 89, 91]
})

print("="*70)
print("ORIGINAL DATASET")
print("="*70)
print(df)
```

```python
# PART 2: SORTING BY SINGLE COLUMN
```

```python
# 2.1 Ascending Order (Default)

print("="*70)
print("SORTING BY SINGLE COLUMN")
print("="*70)

# Sort by Math_Score in ascending order (lowest to highest)
print("\n Sort by Math_Score (Ascending):")
df_sorted_math = df.sort_values('Math_Score')
print(df_sorted_math)
```

```python
# 2.2 Descending Order

print("\n Sort by Math_Score (Descending):")
df_sorted_math_desc = df.sort_values('Math_Score', ascending=False)
print(df_sorted_math_desc)
```

```python
# 2.3 In-Place Sorting (Modify Original DataFrame)

print("\n In-Place Sorting (modifies original):")
df_copy = df.copy()
df_copy.sort_values('Math_Score', inplace=True)
print("Modified DataFrame:")
print(df_copy)
```

```python
# PART 3: SORTING BY MULTIPLE COLUMNS
```

```python
# 3.1 Sorting by Two Columns
print("="*70)
print("SORTING BY MULTIPLE COLUMNS")
print("="*70)

# Sort by Math_Score (descending) then Science_Score (ascending)
print("\n Sort by Math_Score (desc) then Science_Score (asc):")
df_multi = df.sort_values(['Math_Score', 'Science_Score'], 
                          ascending=[False, True])
print(df_multi)
```

```python
# 3.2 Different Sort Orders for Different Columns
print("\n Sort by Math_Score (desc) and English_Score (desc):")
df_multi_diff = df.sort_values(['Math_Score', 'English_Score'], 
                                ascending=[False, False])
print(df_multi_diff)
```

```python
# 3.3 Sorting with Column Names List

print("\n Sort using dynamic column list:")
columns_to_sort = ['Science_Score', 'Math_Score']
df_dynamic = df.sort_values(columns_to_sort, ascending=False)
print(df_dynamic)
```

```python
 # PART 4: SORTING BY INDEX
```

```python
# 4.1 Sorting by Row Index

print("="*70)
print("SORTING BY INDEX")
print("="*70)

# Create shuffled DataFrame
df_shuffled = df.sample(frac=1)  # Random shuffle
print("\n Shuffled DataFrame:")
print(df_shuffled)
```

```python
# Sort by index
print("\n Sort by index (ascending):")
df_index_sorted = df_shuffled.sort_index()
print(df_index_sorted)
```

```python
# Sort by index descending
print("\n Sort by index (descending):")
df_index_desc = df_shuffled.sort_index(ascending=False)
print(df_index_desc)
```

```python
# 4.2 Sorting by Column Index (for DataFrame)

print("\n Sort columns alphabetically:")
df_columns_sorted = df.sort_index(axis=1)
print(df_columns_sorted)
```

```python
# 4.2 Sorting by Column Index (for DataFrame)

print("\n Sort columns alphabetically:")
df_columns_sorted = df.sort_index(axis=0)
print(df_columns_sorted)
```

```python
# PART 5: CUSTOM SORTING ORDERS
```

```python
# 5.1 Using Custom Key Function

print("="*70)
print("CUSTOM SORTING")
print("="*70)

# Create data with grades
grade_data = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Grade': ['A', 'C', 'B+', 'A-', 'B']
})

print("\n Original Grade Data:")
print(grade_data)
```

```python

# Define custom sort order
grade_order = {'A': 1, 'A-': 2, 'B+': 3, 'B': 4, 'C': 5}

print("\n Custom sorting (A, A-, B+, B, C):")
grade_data['Grade_Order'] = grade_data['Grade'].map(grade_order)
grade_data_sorted = grade_data.sort_values('Grade_Order')
print(grade_data_sorted[['Student', 'Grade']])
```

```python
# 5.2 Sorting by String Length

print("\n Sort by name length:")
df['Name_Length'] = df['Student'].str.len()
df_sorted_length = df.sort_values('Name_Length')
print(df_sorted_length[['Student', 'Name_Length']])
```

```python
 # PART 6: SORTING WITH MISSING VALUES
```

```python
print("="*70)
print("SORTING WITH MISSING VALUES")
print("="*70)

# Create data with NaN values
df_nan = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Score': [85, np.nan, 92, 78, np.nan]
})

print("\n Data with Missing Values:")
print(df_nan)
```

```python
# Default: NaN at the end
print("\n NaN at the end (default):")
print(df_nan.sort_values('Score'))
```

```python
# Put NaN at the beginning
print("\n NaN at the beginning:")
print(df_nan.sort_values('Score', na_position='first'))
```

```python
# Put NaN at the end
print("\n NaN at the end:")
print(df_nan.sort_values('Score', na_position='last'))
```
