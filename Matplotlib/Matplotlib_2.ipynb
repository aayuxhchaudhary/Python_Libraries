```python
# What is a Line Plot?

# A Line Plot is a graph where data points are connected using straight line segments.
# Used when data points follow an ordered sequence, such as:
# -Time
# -Distance
# -Temperature
# -Iterations
# -Epochs in ML training
# -Simulation steps
```

```python
# Why Use Line Plot?

# Line plots are best when we want to study:
# 1.Trend
# -Increasing / decreasing movement
# 2.Rate of Change
# -How fast values change
# 3.Continuity
# -Connected observations
# 4.Comparison
# -Two or more variables over same interval
```

```python
# Real World Applications
# 1.Engineering
# -Sensor readings over time
# -Voltage variation
# -CPU load
# 2.Finance
# -Stock prices
# -Revenue growth
# 3.Machine Learning
# -Loss vs epochs
# -Accuracy vs epochs
# 4.Weather
# -Temperature over days
```

```python
# Basic Line Plot Syntax
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [28,30,31,29,32]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
temp = [28,30,31,29,32]

plt.plot(days, temp, marker='o', color='blue')

plt.title("Daily Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.grid(True)

plt.show()
```

```python
# Output Interpretation

# We observe:
# -Day 1 → 28°C
# -Rising till Day 3
# -Drop at Day 4
# -Highest at Day 5
# This is where analysis begins.
```

```python
# Mathematical Understanding
# A line chart visually approximates function behavior.
# y=2x+1
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10,1)
y = 2*x + 1

plt.plot(x,y)
plt.title("Linear Function")
plt.grid(True)
plt.show()
```

```python
# Markers in Line Plot
# Markers highlight actual data points.
plt.plot(x, y, marker='*')
```

```python
# | Marker | Meaning  |
# | ------ | -------- |
# | o      | circle   |
# | s      | square   |
# | ^      | triangle |
# | *      | star     |
# | +      | plus     |
```

```python
# Line Styles
plt.plot(x,y, linestyle='-.')
```

```python
# | Style | Meaning  |
# | ----- | -------- |
# | -     | solid    |
# | --    | dashed   |
# | :     | dotted   |
# | -.    | dash-dot |
```

```python
# Colors
plt.plot(x,y,color='red')
```

```python
# Examples:

# blue:by-default
# red
# green
# black
# orange
```

```python
# Combined Styling Example

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,6,4,8,7]

plt.plot(
    x, y,
    color='green',
    linestyle='--',
    linewidth=3,
    marker='o',
    markersize=8
)

plt.title("Styled Line Plot")
plt.grid(True)
plt.show()
```

```python
# Multiple Lines in One Graph
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.plot(x,y1,label="Series A")
plt.plot(x,y2,label="Series B")

plt.legend(loc="lower right")  # or loc=4
plt.show()
```

```python
# Common loc options:
# Location	String	Code
# Best	'best'	0
# Upper right	'upper right'	1
# Upper left	'upper left'	2
# Lower left	'lower left'	3
# Lower right	'lower right'	4
# Right	'right'	5
# Center left	'center left'	6
# Center right	'center right'	7
# Lower center	'lower center'	8
# Upper center	'upper center'	9
# Center	'center'	10
```

```python
# Why Multiple Lines use?

# Used for comparison:
# -Company A vs Company B sales
# -Train vs Validation accuracy
# -Urban vs Rural population growth
```

```python
# Grid Usage
# Why useful?
# -Easier reading values
# -Better interpretation
# -Research style clarity
plt.grid(True)
```

```python
# Axis Limits:Use when focusing on specific region.

plt.xlim(1,5)
plt.ylim(0,40)
```

```python
# Time Series Line Plot:
# Useful for:
# -temporal analytics
# -business forecasting

months = ["Jan","Feb","Mar","Apr","May"]
sales = [100,120,130,160,150]

plt.plot(months,sales,marker='o')
plt.title("Monthly Sales")
plt.show()
```

```python
# Sine Wave Example:y=sin(x)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Sine Wave")
plt.grid(True)
plt.show()
```

```python
# Trend Analysis Concepts

# We should identify:
# 1.Uptrend
# -values increasing
# 2.Downtrend
# -Values decreasing
# 3.Cyclic Pattern
# -Repeating ups/downs
# 4.Noise
# -Random fluctuations
# 5.Outlier
# -Sudden abnormal spike/drop
```

```python
# Example with Outlier
x = [1,2,3,4,5,6]
y = [10,12,11,13,50,14]

plt.plot(x,y,marker='o')
plt.title("Outlier Detection")
plt.show()

# Que.Why is point 50 unusual?
```

```python
# Annotation on Important Point
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,30,20,18]

plt.plot(x,y,marker='o')

plt.annotate(
    "Peak Value",
    xy=(3,30),
    xytext=(3.2,25),
    arrowprops=dict(facecolor='black')
)

plt.show()
# Useful in reports.
```
